## üìê Linear Algebra Fundamentals

### Vector Spaces and Operations

A **vector space** $V$ over a field $\mathbb{F}$ (usually $\mathbb{R}$ or $\mathbb{C}$) is a set equipped with two operations:

- **Vector addition**: $\mathbf{u} + \mathbf{v} \in V$ for all $\mathbf{u}, \mathbf{v} \in V$
- **Scalar multiplication**: $\alpha \mathbf{v} \in V$ for all $\alpha \in \mathbb{F}$, $\mathbf{v} \in V$

### Inner Products and Norms

**Inner Product**: A function $\langle \cdot, \cdot \rangle: V \times V \rightarrow \mathbb{R}$ that satisfies:

1. **Symmetry**: $\langle \mathbf{x}, \mathbf{y} \rangle = \langle \mathbf{y}, \mathbf{x} \rangle$
2. **Linearity**: $\langle \alpha\mathbf{x} + \beta\mathbf{y}, \mathbf{z} \rangle = \alpha\langle \mathbf{x}, \mathbf{z} \rangle + \beta\langle \mathbf{y}, \mathbf{z} \rangle$
3. **Positive definiteness**: $\langle \mathbf{x}, \mathbf{x} \rangle \geq 0$ with equality iff $\mathbf{x} = \mathbf{0}$

**Standard inner product**: $\langle \mathbf{x}, \mathbf{y} \rangle = \mathbf{x}^T \mathbf{y} = \sum_{i=1}^n x_i y_i$

### Common Norms

1. **$\ell_1$ norm**: $\|\mathbf{x}\|_1 = \sum_{i=1}^n |x_i|$
2. **$\ell_2$ norm (Euclidean)**: $\|\mathbf{x}\|_2 = \sqrt{\sum_{i=1}^n x_i^2}$
3. **$\ell_\infty$ norm**: $\|\mathbf{x}\|_\infty = \max_{i} |x_i|$
4. **$\ell_p$ norm**: $\|\mathbf{x}\|_p = \left(\sum_{i=1}^n |x_i|^p\right)^{1/p}$

### Matrix Operations

**Important matrices for optimization**:

- **Positive Definite**: $\mathbf{A} \succ 0$ if $\mathbf{x}^T\mathbf{A}\mathbf{x} > 0$ for all $\mathbf{x} \neq \mathbf{0}$
- **Positive Semidefinite**: $\mathbf{A} \succeq 0$ if $\mathbf{x}^T\mathbf{A}\mathbf{x} \geq 0$ for all $\mathbf{x}$
- **Symmetric**: $\mathbf{A} = \mathbf{A}^T$

---

## üßÆ Calculus Review

### Gradients and Directional Derivatives

**Gradient**: For $f: \mathbb{R}^n \rightarrow \mathbb{R}$:
$$\nabla f(\mathbf{x}) = \begin{bmatrix} \frac{\partial f}{\partial x_1} \\ \vdots \\ \frac{\partial f}{\partial x_n} \end{bmatrix}$$

**Directional Derivative**: In direction $\mathbf{d}$:
$$D_{\mathbf{d}} f(\mathbf{x}) = \langle \nabla f(\mathbf{x}), \mathbf{d} \rangle$$

### Hessian Matrix

**Hessian**: Matrix of second partial derivatives:

$$
\nabla^2 f(\mathbf{x}) = \mathbf{H}_f(\mathbf{x}) = \begin{bmatrix}
\frac{\partial^2 f}{\partial x_1^2} & \cdots & \frac{\partial^2 f}{\partial x_1 \partial x_n} \\
\vdots & \ddots & \vdots \\
\frac{\partial^2 f}{\partial x_n \partial x_1} & \cdots & \frac{\partial^2 f}{\partial x_n^2}
\end{bmatrix}
$$

### Taylor Series Expansion

**First-order**: $f(\mathbf{x} + \mathbf{h}) \approx f(\mathbf{x}) + \nabla f(\mathbf{x})^T \mathbf{h}$

**Second-order**: $f(\mathbf{x} + \mathbf{h}) \approx f(\mathbf{x}) + \nabla f(\mathbf{x})^T \mathbf{h} + \frac{1}{2}\mathbf{h}^T \nabla^2 f(\mathbf{x}) \mathbf{h}$

---

## üí° Key Insights

1. **Geometric Intuition**: Norms measure "distance" from origin
2. **Optimization Connection**: Gradients point in direction of steepest increase
3. **Curvature Information**: Hessian determines local shape of function
4. **Convexity Preview**: Positive semidefinite Hessian ‚Üí convex function

## üìã General Optimization Problem

An **optimization problem** has the general form:

$$
\begin{align}
\text{minimize} \quad & f_0(\mathbf{x}) \\
\text{subject to} \quad & f_i(\mathbf{x}) \leq 0, \quad i = 1, \ldots, m \\
& h_j(\mathbf{x}) = 0, \quad j = 1, \ldots, p
\end{align}
$$

Where:

- $\mathbf{x} \in \mathbb{R}^n$ is the **optimization variable** (decision variable)
- $f_0: \mathbb{R}^n \rightarrow \mathbb{R}$ is the **objective function** (cost function)
- $f_i: \mathbb{R}^n \rightarrow \mathbb{R}$ are **inequality constraint functions**
- $h_j: \mathbb{R}^n \rightarrow \mathbb{R}$ are **equality constraint functions**

### Key Terminology

- **Feasible set**: $\mathcal{F} = \{\mathbf{x} : f_i(\mathbf{x}) \leq 0, h_j(\mathbf{x}) = 0\}$
- **Feasible point**: $\mathbf{x} \in \mathcal{F}$
- **Optimal value**: $p^* = \inf\{f_0(\mathbf{x}) : \mathbf{x} \in \mathcal{F}\}$
- **Optimal point**: $\mathbf{x}^* \in \mathcal{F}$ such that $f_0(\mathbf{x}^*) = p^*$

---

## üè∑Ô∏è Classification of Optimization Problems

### 1. Linear Programming (LP)

- **Objective**: Linear function
- **Constraints**: Linear functions
- **Form**: $\min \mathbf{c}^T\mathbf{x}$ subject to $\mathbf{A}\mathbf{x} \leq \mathbf{b}$

### 2. Quadratic Programming (QP)

- **Objective**: Quadratic function
- **Constraints**: Linear functions
- **Form**: $\min \frac{1}{2}\mathbf{x}^T\mathbf{Q}\mathbf{x} + \mathbf{c}^T\mathbf{x}$ subject to $\mathbf{A}\mathbf{x} \leq \mathbf{b}$

### 3. Convex Optimization

- **Objective**: Convex function
- **Constraints**: Convex functions (for inequalities)
- **Key Property**: Any local minimum is global minimum

### 4. Nonlinear Programming (NLP)

- **General case**: Nonlinear objective and/or constraints
- **Challenge**: May have multiple local minima

---

## üîç Convex vs. Non-Convex Problems

### Convex Optimization Problem

A problem is **convex** if:

1. Objective function $f_0$ is convex
2. Inequality constraints $f_i$ are convex
3. Equality constraints $h_j$ are affine (linear)

### Why Convex Problems Are Special

1. **Global Optimality**: Any local minimum is a global minimum
2. **Efficient Algorithms**: Polynomial-time algorithms exist
3. **Duality**: Strong duality often holds
4. **Robustness**: Small changes in data lead to small changes in solution

---

## üåü Real-World Examples

### Example 1: Portfolio Optimization

**Problem**: Allocate wealth among assets to minimize risk

$$
\begin{align}
\text{minimize} \quad & \mathbf{x}^T \Sigma \mathbf{x} \\
\text{subject to} \quad & \boldsymbol{\mu}^T \mathbf{x} \geq r_{min} \\
& \mathbf{1}^T \mathbf{x} = 1 \\
& \mathbf{x} \geq 0
\end{align}
$$

### Example 2: Support Vector Machine

**Problem**: Find optimal separating hyperplane

$$
\begin{align}
\text{minimize} \quad & \frac{1}{2}\|\mathbf{w}\|^2 + C\sum_{i=1}^m \xi_i \\
\text{subject to} \quad & y_i(\mathbf{w}^T\mathbf{x}_i + b) \geq 1 - \xi_i \\
& \xi_i \geq 0
\end{align}
$$

### Example 3: Signal Processing

**Problem**: Denoise signal with total variation regularization

$$
\begin{align}
\text{minimize} \quad & \frac{1}{2}\|\mathbf{x} - \mathbf{y}\|_2^2 + \lambda \|\mathbf{D}\mathbf{x}\|_1
\end{align}
$$

---

## üí≠ Geometric Interpretation

Think of optimization as:

1. **Objective**: Defines "height" at each point
2. **Constraints**: Define allowed region (feasible set)
3. **Solution**: Lowest point in allowed region

### Analogy: Mountain Climbing

- **Feasible set**: Safe climbing area
- **Objective function**: Altitude (we want to minimize = go downhill)
- **Constraints**: Boundaries you cannot cross
- **Optimal solution**: Lowest safe point

---

## ‚ö†Ô∏è Common Challenges

1. **Non-convexity**: Multiple local minima
2. **Infeasibility**: No feasible solution exists
3. **Unboundedness**: Objective can be made arbitrarily small
4. **Numerical issues**: Ill-conditioning, numerical precision

## üéØ Key Insights

1. **Problem formulation is crucial**: Same underlying problem can have multiple formulations
2. **Convexity is gold**: Always check if your problem is convex
3. **Constraints shape solutions**: Understanding feasible set geometry is essential
4. **Real problems are messy**: Theory guides practice, but implementation requires care


```python
# Practical Implementation: Optimization Problems

def simple_optimization_example():
    """Solve a simple optimization problem using scipy"""
    # Example: minimize (x-2)¬≤ + (y-1)¬≤ subject to x + y ‚â§ 3, x ‚â• 0, y ‚â• 0
    
    # Objective function
    def objective(vars):
        x, y = vars
        return (x - 2)**2 + (y - 1)**2
    
    # Constraints
    constraints = [
        {'type': 'ineq', 'fun': lambda vars: 3 - vars[0] - vars[1]},  # x + y ‚â§ 3
        {'type': 'ineq', 'fun': lambda vars: vars[0]},                # x ‚â• 0
        {'type': 'ineq', 'fun': lambda vars: vars[1]}                 # y ‚â• 0
    ]
    
    # Initial guess
    x0 = [1, 1]
    
    # Solve
    result = opt.minimize(objective, x0, method='SLSQP', constraints=constraints)
    
    print("Simple Optimization Example:")
    print("Minimize: (x-2)¬≤ + (y-1)¬≤")
    print("Subject to: x + y ‚â§ 3, x ‚â• 0, y ‚â• 0")
    print(f"Solution: x* = {result.x}")
    print(f"Optimal value: f* = {result.fun:.4f}")
    print(f"Success: {result.success}")
    
    # Visualize
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Plot objective function contours and feasible region
    x = np.linspace(-0.5, 3.5, 100)
    y = np.linspace(-0.5, 3.5, 100)
    X, Y = np.meshgrid(x, y)
    Z = (X - 2)**2 + (Y - 1)**2
    
    # Contour plot
    contour = ax1.contour(X, Y, Z, levels=20)
    ax1.clabel(contour, inline=True, fontsize=8)
    
    # Feasible region
    x_feasible = np.array([0, 3, 0, 0])
    y_feasible = np.array([0, 0, 3, 0])
    ax1.fill(x_feasible, y_feasible, alpha=0.3, color='lightblue', label='Feasible Region')
    
    # Optimal point
    ax1.plot(result.x[0], result.x[1], 'ro', markersize=10, label='Optimal Point')
    ax1.plot(2, 1, 'gs', markersize=10, label='Unconstrained Optimum')
    
    ax1.set_xlabel('x')
    ax1.set_ylabel('y')
    ax1.set_title('Constrained Optimization')
    ax1.legend()
    ax1.grid(True)
    ax1.set_xlim(-0.5, 3.5)
    ax1.set_ylim(-0.5, 3.5)
    
    # 3D visualization
    ax2 = fig.add_subplot(122, projection='3d')
    ax2.plot_surface(X, Y, Z, alpha=0.7, cmap='viridis')
    ax2.scatter([result.x[0]], [result.x[1]], [result.fun], color='red', s=100)
    ax2.set_xlabel('x')
    ax2.set_ylabel('y')
    ax2.set_zlabel('f(x,y)')
    ax2.set_title('3D View of Objective Function')
    
    plt.tight_layout()
    plt.show()

def problem_classification_demo():
    """Demonstrate different types of optimization problems"""
    
    print("Problem Classification Examples:\n")
    
    # 1. Linear Programming Example
    print("1. LINEAR PROGRAMMING")
    print("   minimize: 3x‚ÇÅ + 2x‚ÇÇ")
    print("   subject to: x‚ÇÅ + x‚ÇÇ ‚â§ 4")
    print("              2x‚ÇÅ + x‚ÇÇ ‚â§ 6") 
    print("              x‚ÇÅ, x‚ÇÇ ‚â• 0")
    
    # Solve LP using scipy
    c = [3, 2]  # Objective coefficients
    A_ub = [[1, 1], [2, 1]]  # Inequality constraint matrix
    b_ub = [4, 6]  # Inequality constraint bounds
    bounds = [(0, None), (0, None)]  # Variable bounds
    
    result_lp = opt.linprog(c, A_ub=A_ub, b_ub=b_ub, bounds=bounds, method='highs')
    print(f"   Solution: {result_lp.x}")
    print(f"   Optimal value: {result_lp.fun}")
    
    # 2. Quadratic Programming Example  
    print("\n2. QUADRATIC PROGRAMMING")
    print("   minimize: ¬Ωx‚ÇÅ¬≤ + x‚ÇÇ¬≤ - x‚ÇÅx‚ÇÇ - 2x‚ÇÅ - 6x‚ÇÇ")
    print("   subject to: x‚ÇÅ + x‚ÇÇ ‚â§ 2")
    print("              -x‚ÇÅ + 2x‚ÇÇ ‚â§ 2")
    print("              2x‚ÇÅ + x‚ÇÇ ‚â§ 3")
    
    # QP objective: ¬Ωx'Qx + c'x
    Q = np.array([[1, -0.5], [-0.5, 2]])  # Quadratic term
    c = np.array([-2, -6])  # Linear term
    
    def qp_objective(x):
        return 0.5 * x.T @ Q @ x + c.T @ x
    
    constraints_qp = [
        {'type': 'ineq', 'fun': lambda x: 2 - x[0] - x[1]},
        {'type': 'ineq', 'fun': lambda x: 2 + x[0] - 2*x[1]},
        {'type': 'ineq', 'fun': lambda x: 3 - 2*x[0] - x[1]}
    ]
    
    result_qp = opt.minimize(qp_objective, [0, 0], constraints=constraints_qp)
    print(f"   Solution: {result_qp.x}")
    print(f"   Optimal value: {result_qp.fun:.4f}")
    
    # 3. Non-convex Example
    print("\n3. NON-CONVEX PROBLEM")
    print("   minimize: x‚ÇÅ‚Å¥ + x‚ÇÇ‚Å¥ - 4x‚ÇÅ¬≤ - 4x‚ÇÇ¬≤ + x‚ÇÅx‚ÇÇ")
    print("   (Multiple local minima)")
    
    def nonconvex_objective(x):
        return x[0]**4 + x[1]**4 - 4*x[0]**2 - 4*x[1]**2 + x[0]*x[1]
    
    # Try different starting points
    starting_points = [[0, 0], [2, 2], [-2, -2], [2, -2]]
    
    for i, x0 in enumerate(starting_points):
        result_nc = opt.minimize(nonconvex_objective, x0, method='BFGS')
        print(f"   Start {x0}: Solution {result_nc.x}, Value {result_nc.fun:.4f}")

def convex_vs_nonconvex_visualization():
    """Visualize convex vs non-convex functions"""
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))
    
    x = np.linspace(-3, 3, 100)
    
    # Convex function: x¬≤
    y1 = x**2
    ax1.plot(x, y1, 'b-', linewidth=3)
    ax1.set_title('Convex: f(x) = x¬≤', fontsize=14)
    ax1.grid(True)
    ax1.set_ylabel('f(x)')
    
    # Non-convex function: x‚Å¥ - 4x¬≤
    y2 = x**4 - 4*x**2
    ax2.plot(x, y2, 'r-', linewidth=3)
    ax2.set_title('Non-convex: f(x) = x‚Å¥ - 4x¬≤', fontsize=14)
    ax2.grid(True)
    
    # 2D Convex function
    x_2d = np.linspace(-2, 2, 50)
    y_2d = np.linspace(-2, 2, 50)
    X, Y = np.meshgrid(x_2d, y_2d)
    Z_convex = X**2 + Y**2  # Convex
    
    contour1 = ax3.contour(X, Y, Z_convex, levels=15)
    ax3.clabel(contour1, inline=True, fontsize=8)
    ax3.set_title('Convex: f(x,y) = x¬≤ + y¬≤', fontsize=14)
    ax3.set_xlabel('x')
    ax3.set_ylabel('y')
    ax3.grid(True)
    
    # 2D Non-convex function
    Z_nonconvex = X**4 + Y**4 - 4*X**2 - 4*Y**2 + X*Y  # Non-convex
    
    contour2 = ax4.contour(X, Y, Z_nonconvex, levels=15)
    ax4.clabel(contour2, inline=True, fontsize=8)
    ax4.set_title('Non-convex: f(x,y) = x‚Å¥ + y‚Å¥ - 4x¬≤ - 4y¬≤ + xy', fontsize=14)
    ax4.set_xlabel('x')
    ax4.set_ylabel('y')
    ax4.grid(True)
    
    plt.tight_layout()
    plt.show()
    
    print("Key Observations:")
    print("‚Ä¢ Convex functions have a single global minimum")
    print("‚Ä¢ Non-convex functions can have multiple local minima")
    print("‚Ä¢ Convex functions have 'bowl' shape - no local minima that aren't global")
    print("‚Ä¢ Non-convex optimization is generally much harder")

# Run demonstrations
simple_optimization_example()
problem_classification_demo()
convex_vs_nonconvex_visualization()
```

# Section 4: Convex Sets - Definition and Properties

## üéØ Learning Objectives

- Understand key properties and operations on convex sets
- Visualize convex sets in 2D and 3D
- Learn to identify and work with common convex sets

---

## üìê Definition of Convex Sets

A set $C \subseteq \mathbb{R}^n$ is **convex** if for any two points $\mathbf{x}, \mathbf{y} \in C$ and any $\theta \in [0,1]$:

$$\theta \mathbf{x} + (1-\theta) \mathbf{y} \in C$$

### üß† Intuitive Understanding

- **Geometric**: Any line segment between two points in the set lies entirely within the set
- **No "dents"**: Convex sets have no indentations or holes
- **Think**: Soap bubbles, solid balls, half-spaces

### Mathematical Interpretation

The expression $\theta \mathbf{x} + (1-\theta) \mathbf{y}$ represents:

- **Convex combination** of $\mathbf{x}$ and $\mathbf{y}$
- **Weighted average** with weights $\theta$ and $(1-\theta)$
- **Parametric line segment** from $\mathbf{y}$ (when $\theta = 0$) to $\mathbf{x}$ (when $\theta = 1$)

---

## üèóÔ∏è Important Properties of Convex Sets

### 1. **Intersection Property**

The intersection of any collection of convex sets is convex.

**Proof Sketch**: If $\mathbf{x}, \mathbf{y} \in \bigcap_i C_i$, then $\mathbf{x}, \mathbf{y} \in C_i$ for all $i$. Since each $C_i$ is convex, $\theta \mathbf{x} + (1-\theta) \mathbf{y} \in C_i$ for all $i$.

### 2. **Union Property**

The union of convex sets is **generally not convex**.

**Example**: Two disjoint balls - their union is not convex.

### 3. **Affine Transformations**

If $C$ is convex and $f(\mathbf{x}) = \mathbf{A}\mathbf{x} + \mathbf{b}$, then:

- $f(C) = \{\mathbf{A}\mathbf{x} + \mathbf{b} : \mathbf{x} \in C\}$ is convex
- $f^{-1}(C) = \{\mathbf{x} : \mathbf{A}\mathbf{x} + \mathbf{b} \in C\}$ is convex

---

## üìö Examples of Convex Sets

### 1. **Basic Sets**

- **Empty set**: $\emptyset$ (convex by definition)
- **Single point**: $\{\mathbf{x}_0\}$
- **Entire space**: $\mathbb{R}^n$
- **Line segment**: $\{\theta \mathbf{x} + (1-\theta) \mathbf{y} : \theta \in [0,1]\}$

### 2. **Hyperplanes and Half-spaces**

- **Hyperplane**: $\{\mathbf{x} : \mathbf{a}^T \mathbf{x} = b\}$ where $\mathbf{a} \neq \mathbf{0}$
- **Half-space**: $\{\mathbf{x} : \mathbf{a}^T \mathbf{x} \leq b\}$

### 3. **Balls and Ellipsoids**

- **Euclidean ball**: $B(\mathbf{x}_c, r) = \{\mathbf{x} : \|\mathbf{x} - \mathbf{x}_c\|_2 \leq r\}$
- **Ellipsoid**: $\{\mathbf{x} : (\mathbf{x} - \mathbf{x}_c)^T \mathbf{P}^{-1} (\mathbf{x} - \mathbf{x}_c) \leq 1\}$ where $\mathbf{P} \succ 0$

### 4. **Norm Balls**

- **$\ell_p$ ball**: $\{\mathbf{x} : \|\mathbf{x}\|_p \leq 1\}$ for $p \geq 1$
- **Special cases**:
  - $p = 1$: Diamond (cross-polytope)
  - $p = 2$: Circle/sphere
  - $p = \infty$: Square/cube

### 5. **Polyhedra**

- **Definition**: $\{\mathbf{x} : \mathbf{A}\mathbf{x} \leq \mathbf{b}\}$
- **Intersection of half-spaces**
- **Examples**: Simplices, cubes, feasible sets of linear programs

---

## üîß Operations on Convex Sets

### 1. **Convex Hull**

The **convex hull** of a set $S$ is the smallest convex set containing $S$:
$$\text{conv}(S) = \left\{\sum_{i=1}^k \theta_i \mathbf{x}_i : \mathbf{x}_i \in S, \theta_i \geq 0, \sum_{i=1}^k \theta_i = 1\right\}$$

### 2. **Minkowski Sum**

$$C_1 \oplus C_2 = \{\mathbf{x}_1 + \mathbf{x}_2 : \mathbf{x}_1 \in C_1, \mathbf{x}_2 \in C_2\}$$

### 3. **Cartesian Product**

$$C_1 \times C_2 = \{(\mathbf{x}_1, \mathbf{x}_2) : \mathbf{x}_1 \in C_1, \mathbf{x}_2 \in C_2\}$$

---

## üé® Geometric Insights

### Separating Hyperplane Theorem

If $C$ and $D$ are disjoint convex sets, then there exists a hyperplane that separates them:
$$\mathbf{a}^T \mathbf{x} \leq b \leq \mathbf{a}^T \mathbf{y} \quad \text{for all } \mathbf{x} \in C, \mathbf{y} \in D$$

### Supporting Hyperplane Theorem

For any convex set $C$ and boundary point $\mathbf{x}_0 \in \partial C$, there exists a supporting hyperplane:
$$\mathbf{a}^T \mathbf{x} \leq \mathbf{a}^T \mathbf{x}_0 \quad \text{for all } \mathbf{x} \in C$$

---

## üí° Key Insights

1. **Convexity is preserved under intersection** - This is crucial for constraint satisfaction
2. **Linear transformations preserve convexity** - Enables problem reformulation
3. **Convex hull gives "smallest" convex set** - Important for approximation
4. **Separating hyperplanes exist** - Foundation for duality theory

## ‚ö†Ô∏è Common Misconceptions

1. **Union of convex sets is NOT always convex**
2. **Convex ‚â† Connected** (e.g., union of two disjoint points is convex)
3. **Bounded ‚â† Convex** (circles are convex and bounded, but not all bounded sets are convex)
4. **Open vs. Closed**: Both can be convex

---

## üîç Checking Convexity

To verify if a set is convex:

2. **Representation**: Express as intersection of convex sets
3. **Transformation**: Show it's the image/preimage of a known convex set under affine map

# Section 5: Convex Functions - Theory and Characterization

## üìà Definition of Convex Functions

A function $f: \mathbb{R}^n \rightarrow \mathbb{R}$ is **convex** if its domain $\text{dom}(f)$ is a convex set and for all $\mathbf{x}, \mathbf{y} \in \text{dom}(f)$ and $\theta \in [0,1]$:

$$f(\theta \mathbf{x} + (1-\theta) \mathbf{y}) \leq \theta f(\mathbf{x}) + (1-\theta) f(\mathbf{y})$$

### üß† Geometric Interpretation

- **Line segments lie above function**: The function value at any point on a line segment is at most the linear interpolation of the endpoint values
- **"Bowl shape"**: No local minima that aren't global
- **Epigraph is convex**: $\text{epi}(f) = \{(\mathbf{x}, t) : f(\mathbf{x}) \leq t\}$

### Strict Convexity

A function is **strictly convex** if the inequality is strict for $\mathbf{x} \neq \mathbf{y}$:
$$f(\theta \mathbf{x} + (1-\theta) \mathbf{y}) < \theta f(\mathbf{x}) + (1-\theta) f(\mathbf{y})$$

---

## üîç Characterizations of Convexity

### 1. **First-Order Condition** (Differentiable Case)

$f$ is convex if and only if for all $\mathbf{x}, \mathbf{y} \in \text{dom}(f)$:
$$f(\mathbf{y}) \geq f(\mathbf{x}) + \nabla f(\mathbf{x})^T (\mathbf{y} - \mathbf{x})$$

**Interpretation**: The first-order Taylor approximation is a global underestimator.

### 2. **Second-Order Condition** (Twice Differentiable Case)

$f$ is convex if and only if its Hessian is positive semidefinite:
$$\nabla^2 f(\mathbf{x}) \succeq 0 \quad \text{for all } \mathbf{x} \in \text{dom}(f)$$

**For strictly convex**: $\nabla^2 f(\mathbf{x}) \succ 0$

### 3. **Epigraph Characterization**

$f$ is convex if and only if its epigraph is a convex set:
$$\text{epi}(f) = \{(\mathbf{x}, t) : \mathbf{x} \in \text{dom}(f), f(\mathbf{x}) \leq t\}$$

---

## üìö Examples of Convex Functions

### 1. **Affine Functions**

$f(\mathbf{x}) = \mathbf{a}^T \mathbf{x} + b$ (both convex and concave)

### 2. **Quadratic Functions**

$f(\mathbf{x}) = \frac{1}{2}\mathbf{x}^T \mathbf{Q} \mathbf{x} + \mathbf{b}^T \mathbf{x} + c$

- Convex if $\mathbf{Q} \succeq 0$
- Strictly convex if $\mathbf{Q} \succ 0$

### 3. **Norms**

- All norms are convex: $\|\mathbf{x}\|$ for any norm
- Examples: $\|\mathbf{x}\|_1, \|\mathbf{x}\|_2, \|\mathbf{x}\|_\infty$

### 4. **Exponential Functions**

- $e^x$ on $\mathbb{R}$
- $e^{ax}$ for any $a \in \mathbb{R}$

### 5. **Logarithmic Functions**

- $-\log x$ on $(0, \infty)$ (convex)
- $x \log x$ on $(0, \infty)$ (convex)

### 6. **Power Functions**

- $x^p$ on $(0, \infty)$ for $p \geq 1$ or $p \leq 0$
- $|x|^p$ for $p \geq 1$

### 7. **Maximum Function**

$f(\mathbf{x}) = \max\{f_1(\mathbf{x}), \ldots, f_m(\mathbf{x})\}$ where each $f_i$ is convex

---

## üîß Operations Preserving Convexity

### 1. **Nonnegative Weighted Sum**

If $f_1, \ldots, f_m$ are convex and $w_i \geq 0$, then:
$$f(\mathbf{x}) = \sum_{i=1}^m w_i f_i(\mathbf{x})$$
is convex.

### 2. **Pointwise Maximum**

$$f(\mathbf{x}) = \max_{i} f_i(\mathbf{x})$$

### 3. **Composition Rules**

- If $g$ is convex and $h$ is convex nondecreasing, then $h(g(\mathbf{x}))$ is convex
- If $g$ is concave and $h$ is convex nonincreasing, then $h(g(\mathbf{x}))$ is convex

### 4. **Perspective Function**

If $f$ is convex, then $g(x, t) = tf(x/t)$ for $t > 0$ is convex.

---

## üéØ Jensen's Inequality

For a convex function $f$ and $\theta_i \geq 0$, $\sum_i \theta_i = 1$:
$$f\left(\sum_{i=1}^m \theta_i \mathbf{x}_i\right) \leq \sum_{i=1}^m \theta_i f(\mathbf{x}_i)$$

This is a generalization of the basic convexity definition.

---

## üìä Log-Convex Functions

A function $f > 0$ is **log-convex** if $\log f$ is convex.

**Properties**:

- Log-convex ‚áí Convex (but not vice versa)
- Product of log-convex functions is log-convex
- Powers preserve log-convexity: if $f$ is log-convex, so is $f^p$ for $p \geq 0$

---

## üí° Key Insights

1. **Convexity is a global property**: Local behavior determines global behavior
2. **Hessian test is practical**: Check $\nabla^2 f \succeq 0$ to verify convexity
3. **Operations preserve structure**: Build complex convex functions from simple ones
4. **First-order condition**: Gradient gives global underestimator

## ‚ö†Ô∏è Common Mistakes

1. **Domain matters**: Always check that the domain is convex
2. **Composition rules**: Not all compositions preserve convexity
3. **Strict vs. non-strict**: $x^2$ is strictly convex, $|x|$ is convex but not strictly
4. **Multiple variables**: Convexity in each variable doesn't imply joint convexity

```python
# Practical Implementation: Convex Functions

def check_convexity_hessian(func, grad_func, hess_func, test_points):
    """Check convexity using second-order condition"""
    
    print("Convexity Check using Hessian:")
    print("Function is convex if Hessian ‚™∞ 0 everywhere\n")
    
    all_psd = True
    
    for i, point in enumerate(test_points):
        H = hess_func(point)
        eigenvals = np.linalg.eigvals(H)
        min_eigenval = np.min(eigenvals)
        
        is_psd = np.all(eigenvals >= -1e-10)  # Allow small numerical errors
        all_psd = all_psd and is_psd
        
        print(f"Point {i+1}: {point}")
        print(f"  Hessian: \n{H}")
        print(f"  Eigenvalues: {eigenvals}")
        print(f"  Minimum eigenvalue: {min_eigenval:.6f}")
        print(f"  PSD: {'‚úì' if is_psd else '‚úó'}")
        print()
    
    print(f"Overall convexity: {'‚úì CONVEX' if all_psd else '‚úó NOT CONVEX'}")
    return all_psd

def check_convexity_first_order(func, grad_func, test_points, num_checks=50):
    """Check convexity using first-order condition"""
    
    print("Convexity Check using First-Order Condition:")
    print("f(y) ‚â• f(x) + ‚àáf(x)·µÄ(y-x) should hold for all x, y\n")
    
    violations = 0
    
    for i in range(num_checks):
        # Random pair of points
        x = test_points[np.random.randint(len(test_points))]
        y = test_points[np.random.randint(len(test_points))]
        
        fx = func(x)
        fy = func(y)
        grad_x = grad_func(x)
        
        # First-order condition
        linear_approx = fx + np.dot(grad_x, y - x)
        condition_satisfied = fy >= linear_approx - 1e-10
        
        if not condition_satisfied:
            violations += 1
            print(f"Violation {violations}: x={x}, y={y}")
            print(f"  f(y) = {fy:.6f}")
            print(f"  f(x) + ‚àáf(x)·µÄ(y-x) = {linear_approx:.6f}")
            print(f"  Difference: {fy - linear_approx:.6f}")
    
    print(f"\nChecked {num_checks} random pairs")
    print(f"Violations: {violations}")
    print(f"First-order test: {'‚úì PASSED' if violations == 0 else '‚úó FAILED'}")
    
    return violations == 0

def visualize_convex_functions():
    """Visualize convex and non-convex functions"""
    
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    x = np.linspace(-3, 3, 1000)
    
    # Row 1: Convex functions
    functions_convex = [
        (lambda t: t**2, "f(x) = x¬≤", "Quadratic"),
        (lambda t: np.exp(t), "f(x) = eÀ£", "Exponential"), 
        (lambda t: np.abs(t), "f(x) = |x|", "Absolute Value")
    ]
    
    # Row 2: Non-convex functions  
    functions_nonconvex = [
        (lambda t: t**3, "f(x) = x¬≥", "Cubic"),
        (lambda t: np.sin(t), "f(x) = sin(x)", "Sine"),
        (lambda t: t**4 - 2*t**2, "f(x) = x‚Å¥ - 2x¬≤", "Double Well")
    ]
    
    all_functions = [functions_convex, functions_nonconvex]
    titles = ["Convex Functions", "Non-Convex Functions"]
    
    for row, (functions, row_title) in enumerate(zip(all_functions, titles)):
        for col, (func, formula, name) in enumerate(functions):
            ax = axes[row, col]
            
            y = func(x)
            ax.plot(x, y, 'b-', linewidth=3, label=formula)
            
            # Add tangent line for visualization
            x0 = 0 if name != "Absolute Value" else 1  # Avoid non-differentiable point
            if name == "Sine":
                x0 = np.pi/4
            
            # Numerical derivative
            h = 1e-6
            if name != "Absolute Value" or x0 != 0:
                derivative = (func(x0 + h) - func(x0 - h)) / (2 * h)
                y0 = func(x0)
                
                # Tangent line
                tangent_x = x
                tangent_y = y0 + derivative * (tangent_x - x0)
                ax.plot(tangent_x, tangent_y, 'r--', alpha=0.7, label='Tangent')
                ax.plot(x0, y0, 'ro', markersize=8)
            
            ax.set_title(f"{name}\n{formula}")
            ax.grid(True, alpha=0.3)
            ax.legend()
            ax.set_xlabel('x')
            ax.set_ylabel('f(x)')
            
            if row == 0:
                ax.set_ylim(bottom=min(y) - 0.5)
    
    plt.suptitle("Convex vs Non-Convex Functions", fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.show()

def demonstrate_convex_combinations_functions():
    """Show that convex functions satisfy Jensen's inequality"""
    
    # Example: f(x) = x¬≤
    def f(x):
        return x**2
    
    # Two points
    x1, x2 = -2, 3
    
    # Range of theta values
    theta_vals = np.linspace(0, 1, 11)
    
    print("Jensen's Inequality Demonstration:")
    print(f"Function: f(x) = x¬≤")
    print(f"Points: x‚ÇÅ = {x1}, x‚ÇÇ = {x2}")
    print(f"f(x‚ÇÅ) = {f(x1)}, f(x‚ÇÇ) = {f(x2)}")
    print()
    print("Œ∏    | Œ∏x‚ÇÅ+(1-Œ∏)x‚ÇÇ | f(Œ∏x‚ÇÅ+(1-Œ∏)x‚ÇÇ) | Œ∏f(x‚ÇÅ)+(1-Œ∏)f(x‚ÇÇ) | Difference")
    print("-" * 75)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Plot function
    x_plot = np.linspace(-3, 4, 1000)
    y_plot = f(x_plot)
    ax1.plot(x_plot, y_plot, 'b-', linewidth=2, label='f(x) = x¬≤')
    
    # Plot convex combinations
    convex_combo_points = []
    function_values = []
    linear_combo_values = []
    
    for theta in theta_vals:
        # Convex combination of points
        x_combo = theta * x1 + (1 - theta) * x2
        # Function value at convex combination
        f_combo = f(x_combo)
        # Linear combination of function values
        linear_combo = theta * f(x1) + (1 - theta) * f(x2)
        
        convex_combo_points.append(x_combo)
        function_values.append(f_combo)
        linear_combo_values.append(linear_combo)
        
        print(f"{theta:.1f}  | {x_combo:11.1f} | {f_combo:14.1f} | {linear_combo:17.1f} | {linear_combo - f_combo:10.1f}")
    
    # Plot on function
    ax1.scatter(convex_combo_points, function_values, c='red', s=80, zorder=5, label='f(Œ∏x‚ÇÅ+(1-Œ∏)x‚ÇÇ)')
    ax1.scatter([x1, x2], [f(x1), f(x2)], c='green', s=100, zorder=5, marker='s', label='Endpoints')
    
    # Plot linear interpolation
    ax1.plot([x1, x2], [f(x1), f(x2)], 'g--', linewidth=2, alpha=0.7)
    ax1.scatter(convex_combo_points, linear_combo_values, c='orange', s=60, zorder=5, marker='^', label='Œ∏f(x‚ÇÅ)+(1-Œ∏)f(x‚ÇÇ)')
    
    ax1.set_xlabel('x')
    ax1.set_ylabel('f(x)')
    ax1.set_title('Jensen\'s Inequality Visualization')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # Plot the inequality gap
    theta_fine = np.linspace(0, 1, 100)
    gap_values = []
    
    for theta in theta_fine:
        x_combo = theta * x1 + (1 - theta) * x2
        f_combo = f(x_combo)
        linear_combo = theta * f(x1) + (1 - theta) * f(x2)
        gap_values.append(linear_combo - f_combo)
    
    ax2.plot(theta_fine, gap_values, 'purple', linewidth=3)
    ax2.fill_between(theta_fine, 0, gap_values, alpha=0.3, color='purple')
    ax2.axhline(y=0, color='black', linestyle='-', alpha=0.3)
    ax2.set_xlabel('Œ∏')
    ax2.set_ylabel('Œ∏f(x‚ÇÅ) + (1-Œ∏)f(x‚ÇÇ) - f(Œ∏x‚ÇÅ + (1-Œ∏)x‚ÇÇ)')
    ax2.set_title('Convexity Gap (Always ‚â• 0)')
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

def practical_convexity_tests():
    """Test convexity of various functions"""
    
    # Example 1: Quadratic function f(x,y) = x¬≤ + y¬≤ 
    print("=" * 60)
    print("EXAMPLE 1: f(x,y) = x¬≤ + y¬≤")
    print("=" * 60)
    
    def f1(x):
        return x[0]**2 + x[1]**2
    
    def grad_f1(x):
        return np.array([2*x[0], 2*x[1]])
    
    def hess_f1(x):
        return np.array([[2, 0], [0, 2]])
    
    test_points_1 = [np.array([0, 0]), np.array([1, 1]), np.array([-1, 2]), np.array([2, -1])]
    
    check_convexity_hessian(f1, grad_f1, hess_f1, test_points_1)
    print()
    check_convexity_first_order(f1, grad_f1, test_points_1)
    
    # Example 2: Non-convex function f(x,y) = x¬≤ - y¬≤
    print("\n" + "=" * 60)
    print("EXAMPLE 2: f(x,y) = x¬≤ - y¬≤ (Saddle function)")
    print("=" * 60)
    
    def f2(x):
        return x[0]**2 - x[1]**2
    
    def grad_f2(x):
        return np.array([2*x[0], -2*x[1]])
    
    def hess_f2(x):
        return np.array([[2, 0], [0, -2]])
    
    test_points_2 = [np.array([0, 0]), np.array([1, 0]), np.array([0, 1]), np.array([1, 1])]
    
    check_convexity_hessian(f2, grad_f2, hess_f2, test_points_2)
    
# Run demonstrations
visualize_convex_functions()
demonstrate_convex_combinations_functions()
practical_convexity_tests()
```

## üìã Standard Form Convex Problem

A **convex optimization problem** has the form:

$$
\begin{align}
\text{minimize} \quad & f_0(\mathbf{x}) \\
\text{subject to} \quad & f_i(\mathbf{x}) \leq 0, \quad i = 1, \ldots, m \\
& \mathbf{A}\mathbf{x} = \mathbf{b}
\end{align}
$$

Where:

- $f_0, f_1, \ldots, f_m$ are **convex functions**
- Equality constraints are **affine** (linear)
- The problem is **feasible**

### Key Properties

1. **Feasible set is convex**: Intersection of convex sets
2. **Any local optimum is global**: Fundamental property
3. **Efficient algorithms exist**: Polynomial-time solvable

---

## üèóÔ∏è Problem Classes

### 1. **Linear Program (LP)**

$$
\begin{align}
\text{minimize} \quad & \mathbf{c}^T \mathbf{x} \\
\text{subject to} \quad & \mathbf{A} \mathbf{x} \leq \mathbf{b} \\
& \mathbf{C} \mathbf{x} = \mathbf{d}
\end{align}
$$

**Standard form**: $\min \mathbf{c}^T \mathbf{x}$ subject to $\mathbf{A} \mathbf{x} = \mathbf{b}$, $\mathbf{x} \geq 0$

### 2. **Quadratic Program (QP)**

$$
\begin{align}
\text{minimize} \quad & \frac{1}{2} \mathbf{x}^T \mathbf{P} \mathbf{x} + \mathbf{q}^T \mathbf{x} + r \\
\text{subject to} \quad & \mathbf{A} \mathbf{x} \leq \mathbf{b} \\
& \mathbf{C} \mathbf{x} = \mathbf{d}
\end{align}
$$

Where $\mathbf{P} \succeq 0$ for convexity.

### 3. **Quadratically Constrained QP (QCQP)**

$$
\begin{align}
\text{minimize} \quad & \frac{1}{2} \mathbf{x}^T \mathbf{P}_0 \mathbf{x} + \mathbf{q}_0^T \mathbf{x} + r_0 \\
\text{subject to} \quad & \frac{1}{2} \mathbf{x}^T \mathbf{P}_i \mathbf{x} + \mathbf{q}_i^T \mathbf{x} + r_i \leq 0, \quad i = 1, \ldots, m \\
& \mathbf{A} \mathbf{x} = \mathbf{b}
\end{align}
$$

Where $\mathbf{P}_0, \mathbf{P}_1, \ldots, \mathbf{P}_m \succeq 0$.

### 4. **Second-Order Cone Program (SOCP)**

$$
\begin{align}
\text{minimize} \quad & \mathbf{f}^T \mathbf{x} \\
\text{subject to} \quad & \|\mathbf{A}_i \mathbf{x} + \mathbf{b}_i\|_2 \leq \mathbf{c}_i^T \mathbf{x} + d_i, \quad i = 1, \ldots, m \\
& \mathbf{F} \mathbf{x} = \mathbf{g}
\end{align}
$$

### 5. **Semidefinite Program (SDP)**

$$
\begin{align}
\text{minimize} \quad & \mathbf{c}^T \mathbf{x} \\
\text{subject to} \quad & x_1 \mathbf{F}_1 + \cdots + x_n \mathbf{F}_n + \mathbf{G} \preceq 0 \\
& \mathbf{A} \mathbf{x} = \mathbf{b}
\end{align}
$$

Where $\mathbf{F}_i, \mathbf{G}$ are symmetric matrices.

---

## üîÑ Problem Transformations

### 1. **Eliminating Equality Constraints**

If $\mathbf{A} \mathbf{x} = \mathbf{b}$ and $\text{rank}(\mathbf{A}) = p < n$:

1. Find particular solution: $\mathbf{x}_0$ such that $\mathbf{A} \mathbf{x}_0 = \mathbf{b}$
2. Find null space basis: $\mathbf{A} \mathbf{N} = 0$
3. Substitute: $\mathbf{x} = \mathbf{x}_0 + \mathbf{N} \mathbf{z}$

### 2. **Introducing Slack Variables**

Transform $f_i(\mathbf{x}) \leq 0$ to:

- Add variable $s_i \geq 0$
- Replace with $f_i(\mathbf{x}) + s_i = 0$, $s_i \geq 0$

### 3. **Epigraph Form**

Transform $\min f_0(\mathbf{x})$ to:

$$
\begin{align}
\text{minimize} \quad & t \\
\text{subject to} \quad & f_0(\mathbf{x}) \leq t \\
& f_i(\mathbf{x}) \leq 0, \quad i = 1, \ldots, m
\end{align}
$$

---

## üéØ Common Reformulations

### 1. **$\ell_1$ Norm Minimization**

$$\min \|\mathbf{A} \mathbf{x} - \mathbf{b}\|_1$$

**Reformulation**:

$$
\begin{align}
\text{minimize} \quad & \mathbf{1}^T \mathbf{t} \\
\text{subject to} \quad & -\mathbf{t} \leq \mathbf{A} \mathbf{x} - \mathbf{b} \leq \mathbf{t}
\end{align}
$$

### 2. **$\ell_\infty$ Norm Minimization**

$$\min \|\mathbf{A} \mathbf{x} - \mathbf{b}\|_\infty$$

**Reformulation**:

$$
\begin{align}
\text{minimize} \quad & t \\
\text{subject to} \quad & \|\mathbf{A} \mathbf{x} - \mathbf{b}\|_\infty \leq t \\
& -t \mathbf{1} \leq \mathbf{A} \mathbf{x} - \mathbf{b} \leq t \mathbf{1}
\end{align}
$$

### 3. **Robust Linear Programming**

Uncertain LP: $\min \mathbf{c}^T \mathbf{x}$ subject to $\mathbf{a}_i^T \mathbf{x} \leq b_i$ for all $\mathbf{a}_i \in \mathcal{U}_i$

If $\mathcal{U}_i = \{\bar{\mathbf{a}}_i + \mathbf{P}_i \mathbf{u} : \|\mathbf{u}\|_2 \leq 1\}$:

**Reformulation**:

$$
\begin{align}
\text{minimize} \quad & \mathbf{c}^T \mathbf{x} \\
\text{subject to} \quad & \bar{\mathbf{a}}_i^T \mathbf{x} + \|\mathbf{P}_i^T \mathbf{x}\|_2 \leq b_i, \quad i = 1, \ldots, m
\end{align}
$$

---

## üîç Optimality Conditions

### 1. **Unconstrained Case**

For unconstrained $\min f(\mathbf{x})$:

- **Necessary**: $\nabla f(\mathbf{x}^*) = 0$
- **Sufficient**: $\nabla f(\mathbf{x}^*) = 0$ and $\nabla^2 f(\mathbf{x}^*) \succ 0$

### 2. **Equality Constrained Case**

For $\min f(\mathbf{x})$ subject to $\mathbf{A} \mathbf{x} = \mathbf{b}$:

**Lagrangian**: $L(\mathbf{x}, \boldsymbol{\nu}) = f(\mathbf{x}) + \boldsymbol{\nu}^T (\mathbf{A} \mathbf{x} - \mathbf{b})$

**Optimality**: $\nabla_{\mathbf{x}} L(\mathbf{x}^*, \boldsymbol{\nu}^*) = 0$ and $\mathbf{A} \mathbf{x}^* = \mathbf{b}$

### 3. **General Constrained Case**

**KKT Conditions** (necessary for constrained optimum):

1. **Stationarity**: $\nabla f_0(\mathbf{x}^*) + \sum_{i=1}^m \lambda_i^* \nabla f_i(\mathbf{x}^*) + \mathbf{A}^T \boldsymbol{\nu}^* = 0$
2. **Primal feasibility**: $f_i(\mathbf{x}^*) \leq 0$, $\mathbf{A} \mathbf{x}^* = \mathbf{b}$
3. **Dual feasibility**: $\lambda_i^* \geq 0$
4. **Complementary slackness**: $\lambda_i^* f_i(\mathbf{x}^*) = 0$

---

## üí° Modeling Tips

1. **Start simple**: Begin with basic formulation, then add complexity
2. **Check convexity**: Verify all functions are convex
3. **Use standard forms**: Transform to well-known problem classes
4. **Exploit structure**: Identify special structure for efficient algorithms
5. **Validate numerically**: Test on small examples first

## ‚ö†Ô∏è Common Formulation Errors

1. **Non-convex constraints**: $f_i(\mathbf{x}) \geq 0$ where $f_i$ is convex
2. **Non-affine equalities**: $g(\mathbf{x}) = 0$ where $g$ is not affine
3. **Objective direction**: Maximizing convex function (minimize concave instead)
4. **Domain issues**: Forgetting to specify variable domains


```python
# Practical Implementation: Problem Formulation

def solve_portfolio_optimization():
    """Solve a portfolio optimization problem (Markowitz model)"""
    
    # Problem: minimize portfolio risk subject to return constraint
    # min (1/2) x^T Œ£ x
    # s.t. Œº^T x ‚â• r_min
    #      1^T x = 1
    #      x ‚â• 0
    
    print("PORTFOLIO OPTIMIZATION EXAMPLE")
    print("=" * 50)
    
    # Generate synthetic data
    np.random.seed(42)
    n_assets = 4
    
    # Expected returns
    mu = np.array([0.12, 0.10, 0.07, 0.03])
    
    # Covariance matrix (must be positive definite)
    A = np.random.randn(n_assets, n_assets)
    Sigma = A.T @ A / 100  # Ensure positive definite
    
    print(f"Expected returns: {mu}")
    print(f"Covariance matrix:\n{Sigma}")
    
    # Minimum required return
    r_min = 0.08
    
    # Solve using scipy
    from scipy.optimize import minimize
    
    # Objective function: portfolio variance
    def portfolio_variance(x):
        return 0.5 * x.T @ Sigma @ x
    
    # Constraints
    constraints = [
        {'type': 'ineq', 'fun': lambda x: mu.T @ x - r_min},  # Return constraint
        {'type': 'eq', 'fun': lambda x: np.sum(x) - 1}        # Budget constraint
    ]
    
    # Bounds: all weights non-negative
    bounds = [(0, 1) for _ in range(n_assets)]
    
    # Initial guess
    x0 = np.ones(n_assets) / n_assets
    
    # Solve
    result = minimize(portfolio_variance, x0, method='SLSQP', 
                     bounds=bounds, constraints=constraints)
    
    print(f"\nOptimal portfolio:")
    print(f"Weights: {result.x}")
    print(f"Expected return: {mu.T @ result.x:.4f}")
    print(f"Portfolio risk (std): {np.sqrt(2 * result.fun):.4f}")
    print(f"Success: {result.success}")
    
    # Efficient frontier
    returns = np.linspace(0.03, 0.12, 50)
    risks = []
    weights_history = []
    
    for target_return in returns:
        constraints_ef = [
            {'type': 'ineq', 'fun': lambda x, r=target_return: mu.T @ x - r},
            {'type': 'eq', 'fun': lambda x: np.sum(x) - 1}
        ]
        
        result_ef = minimize(portfolio_variance, x0, method='SLSQP',
                           bounds=bounds, constraints=constraints_ef)
        
        if result_ef.success:
            risks.append(np.sqrt(2 * result_ef.fun))
            weights_history.append(result_ef.x)
        else:
            risks.append(np.nan)
            weights_history.append(np.full(n_assets, np.nan))
    
    # Plot efficient frontier
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Efficient frontier
    valid_idx = ~np.isnan(risks)
    ax1.plot(np.array(risks)[valid_idx], np.array(returns)[valid_idx], 'b-', linewidth=2)
    ax1.scatter(np.sqrt(2 * result.fun), mu.T @ result.x, color='red', s=100, zorder=5)
    ax1.set_xlabel('Risk (Standard Deviation)')
    ax1.set_ylabel('Expected Return')
    ax1.set_title('Efficient Frontier')
    ax1.grid(True, alpha=0.3)
    
    # Weight evolution
    weights_array = np.array(weights_history)
    asset_names = ['Asset 1', 'Asset 2', 'Asset 3', 'Asset 4']
    
    for i in range(n_assets):
        ax2.plot(returns, weights_array[:, i], label=asset_names[i], linewidth=2)
    
    ax2.set_xlabel('Target Return')
    ax2.set_ylabel('Portfolio Weight')
    ax2.set_title('Optimal Weights vs Target Return')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

def solve_lasso_regression():
    """Solve LASSO regression using reformulation"""
    
    print("\nLASSO REGRESSION EXAMPLE")
    print("=" * 50)
    
    # Generate synthetic data
    np.random.seed(123)
    m, n = 50, 20  # 50 samples, 20 features
    
    X = np.random.randn(m, n)
    true_beta = np.zeros(n)
    true_beta[:5] = np.random.randn(5)  # Only first 5 features are relevant
    y = X @ true_beta + 0.1 * np.random.randn(m)  # Add noise
    
    print(f"Data: {m} samples, {n} features")
    print(f"True non-zero coefficients: {np.sum(true_beta != 0)}")
    
    # LASSO problem: min (1/2)||XŒ≤ - y||¬≤ + Œª||Œ≤||‚ÇÅ
    # Reformulation as LP:
    # min (1/2)||XŒ≤ - y||¬≤ + Œª(1^T u)
    # s.t. -u ‚â§ Œ≤ ‚â§ u, u ‚â• 0
    
    lambda_reg = 0.1
    
    # Solve using scipy (simple approach)
    def lasso_objective(beta):
        residual = X @ beta - y
        return 0.5 * np.sum(residual**2) + lambda_reg * np.sum(np.abs(beta))    # Use L-BFGS-B which handles bounds\n    result = opt.minimize(lasso_objective, np.zeros(n), method='L-BFGS-B')\n    \n    beta_lasso = result.x\n    \n    print(f\"\\nLASSO Results (Œª = {lambda_reg}):\")\n    print(f\"Non-zero coefficients: {np.sum(np.abs(beta_lasso) > 1e-4)}\")\n    print(f\"Objective value: {result.fun:.4f}\")\n    \n    # Compare with scikit-learn\n    try:\n        from sklearn.linear_model import Lasso\n        lasso_sklearn = Lasso(alpha=lambda_reg / m, max_iter=10000)\n        lasso_sklearn.fit(X, y)\n        beta_sklearn = lasso_sklearn.coef_\n        \n        print(f\"Sklearn non-zero coefficients: {np.sum(np.abs(beta_sklearn) > 1e-4)}\")\n    except ImportError:\n        beta_sklearn = beta_lasso\n        print(\"Scikit-learn not available for comparison\")\n    \n    # Visualization\n    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 5))\n    \n    # True vs estimated coefficients\n    ax1.stem(range(n), true_beta, basefmt=\" \", label='True Œ≤', linefmt='b-', markerfmt='bo')\n    ax1.stem(range(n), beta_lasso, basefmt=\" \", label='LASSO Œ≤', linefmt='r-', markerfmt='ro', alpha=0.7)\n    ax1.set_xlabel('Feature Index')\n    ax1.set_ylabel('Coefficient Value')\n    ax1.set_title('True vs LASSO Coefficients')\n    ax1.legend()\n    ax1.grid(True, alpha=0.3)\n    \n    # Regularization path\n    lambdas = np.logspace(-3, 1, 50)\n    path_coeffs = []\n    \n    for lam in lambdas:\n        def obj(beta):\n            residual = X @ beta - y\n            return 0.5 * np.sum(residual**2) + lam * np.sum(np.abs(beta))\n        \n        res = opt.minimize(obj, np.zeros(n), method='L-BFGS-B')\n        path_coeffs.append(res.x)\n    \n    path_coeffs = np.array(path_coeffs)\n    \n    for i in range(min(10, n)):  # Plot first 10 coefficients\n        ax2.plot(lambdas, path_coeffs[:, i], label=f'Œ≤{i}' if i < 5 else None)\n    \n    ax2.axvline(lambda_reg, color='black', linestyle='--', alpha=0.7, label=f'Used Œª = {lambda_reg}')\n    ax2.set_xlabel('Œª (Regularization Parameter)')\n    ax2.set_ylabel('Coefficient Value')\n    ax2.set_title('LASSO Regularization Path')\n    ax2.set_xscale('log')\n    ax2.legend()\n    ax2.grid(True, alpha=0.3)\n    \n    # Prediction comparison\n    y_pred_lasso = X @ beta_lasso\n    ax3.scatter(y, y_pred_lasso, alpha=0.6)\n    ax3.plot([y.min(), y.max()], [y.min(), y.max()], 'r--', linewidth=2)\n    ax3.set_xlabel('True y')\n    ax3.set_ylabel('Predicted y')\n    ax3.set_title('LASSO Predictions')\n    ax3.grid(True, alpha=0.3)\n    \n    # Add R¬≤ score\n    r2 = 1 - np.sum((y - y_pred_lasso)**2) / np.sum((y - np.mean(y))**2)\n    ax3.text(0.05, 0.95, f'R¬≤ = {r2:.3f}', transform=ax3.transAxes, \n             bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))\n    \n    plt.tight_layout()\n    plt.show()\n\ndef demonstrate_problem_transformations():\n    \"\"\"Show common problem transformations\"\"\"\n    \n    print(\"\\nPROBLEM TRANSFORMATION EXAMPLES\")\n    print(\"=\" * 50)\n    \n    # Example 1: Convert absolute value to linear constraints\n    print(\"\\n1. L1 NORM MINIMIZATION\")\n    print(\"Original: min ||Ax - b||‚ÇÅ\")\n    print(\"Reformulation:\")\n    print(\"  min 1^T t\")\n    print(\"  s.t. -t ‚â§ Ax - b ‚â§ t\")\n    \n    # Generate example\n    np.random.seed(42)\n    A = np.random.randn(10, 5)\n    b = np.random.randn(10)\n    \n    # Original problem (using scipy's minimize with L1 penalty)\n    def l1_objective(x):\n        return np.sum(np.abs(A @ x - b))\n    \n    result_orig = opt.minimize(l1_objective, np.zeros(5), method='L-BFGS-B')\n    \n    # Reformulated problem\n    # Variables: [x1, ..., x5, t1, ..., t10]\n    n_vars = A.shape[1] + A.shape[0]  # x + t variables\n    \n    def reformed_objective(vars):\n        x = vars[:A.shape[1]]\n        t = vars[A.shape[1]:]\n        return np.sum(t)\n    \n    def constraint_lower(vars):\n        x = vars[:A.shape[1]]\n        t = vars[A.shape[1]:]\n        return t + (A @ x - b)  # t + (Ax - b) ‚â• 0 ‚Üí -t ‚â§ Ax - b\n    \n    def constraint_upper(vars):\n        x = vars[:A.shape[1]]\n        t = vars[A.shape[1]:]\n        return t - (A @ x - b)  # t - (Ax - b) ‚â• 0 ‚Üí Ax - b ‚â§ t\n    \n    constraints_reformed = [\n        {'type': 'ineq', 'fun': constraint_lower},\n        {'type': 'ineq', 'fun': constraint_upper},\n        {'type': 'ineq', 'fun': lambda vars: vars[A.shape[1]:]}  # t ‚â• 0\n    ]\n    \n    result_reformed = opt.minimize(reformed_objective, np.zeros(n_vars), \n                                 method='SLSQP', constraints=constraints_reformed)\n    \n    print(f\"Original objective: {result_orig.fun:.6f}\")\n    print(f\"Reformed objective: {result_reformed.fun:.6f}\")\n    print(f\"Difference: {abs(result_orig.fun - result_reformed.fun):.2e}\")\n    \n    # Example 2: Epigraph form\n    print(\"\\n2. EPIGRAPH FORM\")\n    print(\"Original: min f(x)\")\n    print(\"Epigraph: min t s.t. f(x) ‚â§ t\")\n    \n    # Example: min x¬≤\n    def quadratic_obj(x):\n        return x[0]**2\n    \n    # Original\n    result_quad_orig = opt.minimize(quadratic_obj, [1], method='BFGS')\n    \n    # Epigraph form: min t s.t. x¬≤ ‚â§ t\n    def epigraph_obj(vars):\n        # vars = [x, t]\n        return vars[1]  # minimize t\n    \n    def epigraph_constraint(vars):\n        x, t = vars\n        return t - x**2  # t - x¬≤ ‚â• 0\n    \n    constraints_epi = [{'type': 'ineq', 'fun': epigraph_constraint}]\n    \n    result_epi = opt.minimize(epigraph_obj, [1, 2], method='SLSQP', \n                            constraints=constraints_epi)\n    \n    print(f\"Original minimum: x* = {result_quad_orig.x[0]:.6f}, f* = {result_quad_orig.fun:.6f}\")\n    print(f\"Epigraph minimum: x* = {result_epi.x[0]:.6f}, t* = {result_epi.x[1]:.6f}\")\n    print(f\"Both should give same result: {np.allclose([result_quad_orig.x[0], result_quad_orig.fun], result_epi.x, atol=1e-5)}\")\n\n# Run demonstrations\nsolve_portfolio_optimization()\nsolve_lasso_regression()\ndemonstrate_problem_transformations()
solve_portfolio_optimization()
```
