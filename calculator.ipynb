```python
import sympy as sp
import markdown as md
```

```python
def derivitive_functiion(f, variable='x'):
    f = sp.sympify(f)
    derivative = sp.diff(f, variable)
    return derivative
```

```python
def integral_function(f,variable='x'):
    #f = sp.sympify(f)
    integral = sp.integrate(f)
    return integral
```

```python
def solver(equation, variable):
    variable = sp.symbols(variable)
    equation = sp.sympify(equation)
    solutions = sp.solve(equation, variable)
    return solutions
```

```python
def limit_function(f, variable, point):
    variable = sp.symbols(variable)
    f = sp.sympify(f)
    limit = sp.limit(f, variable, point)
    return limit
```

```python
inp = input("Enter a mathematical expression: ").replace("\\", "").replace(" ", "").replace("_", "")
inp
```

```python
if inp.endswith("=0"):
    equation = inp[:-2]
    solutions = solver(equation, 'x')
    print(f"The solutions are: {solutions}")
if 'integrate' in inp:
    func = inp[inp.index('(')+1:inp.index(')')]
    integral = (integral_function(func,'x'))
    print("The integral is:")
    print(integral)
if 'derivate' in inp:
    func = inp[inp.index('(')+1:inp.index(')')]
    derivitive = (derivitive_functiion(func))
    print("The derivative is:")
    print(derivitive)
if 'lim' in inp:
    variable = inp[inp.index('{')+1:inp.index('to')].strip()
    point = float(inp[inp.index('to')+2:inp.index('}')].strip())
    func = inp[inp.index('(')+1:-1]  # Remove the trailing ')'
    func = func.replace('^', '**')  # Replace ^ with ** for Python syntax
    print(variable, point, func)
    defined_limit = limit_function(func, variable, point)
    print(f"The limit is: {defined_limit:.2f}")
```
