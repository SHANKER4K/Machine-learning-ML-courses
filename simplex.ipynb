```python
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
def add_slucky_variables(A, b):
    
    m, n = A.shape
    slack_vars = np.eye(m,m+1)
    A_slack = np.hstack((A, slack_vars,b.reshape(-1,1)))
    return A_slack
```

```python
def simplex(x):
    x = x.astype(float)
    while True:
        min_z = np.min(x[-1, :-1])
        if min_z >= 0:
            break  # Optimal reached
        index_z = np.argmin(x[-1, :-1])
        selected_col = x[:, index_z]
        with np.errstate(divide='ignore', invalid='ignore'):
            ratios = np.where(selected_col > 0, x[:, -1] / selected_col, np.inf)
        ratios = ratios[:-1]  # Exclude last row (objective)
        if np.all(ratios == np.inf):
            raise Exception("Unbounded solution")
        index_r = np.argmin(ratios)
        pivot = x[index_r, index_z]
        x[index_r, :] = x[index_r, :] / pivot
        for i in range(x.shape[0]):
            if i != index_r:
                x[i, :] = x[i, :] - x[i, index_z] * x[index_r, :]
        x = np.round(x, 6)
    return x
```

```python
d = np.array([[1, 1], [4, 1],[-20,-10]])
b = np.array([40, 100,0])
print("Simplex module loaded.")
x = add_slucky_variables(d, b)
x
```

```python
for row in simplex(x):
    print(row)
```
