## üîß Utility Functions

Let's create some utility functions that we'll use throughout the course for visualization and analysis:


````python
def plot_decision_boundary(X, y, classifier, title="SVM Decision Boundary", 
                          feature_names=None, class_names=None, support_vectors=None):
    """
    Plot decision boundary for 2D classification problems
    
    Parameters:
    -----------
    X : array-like, shape (n_samples, 2)
        Feature matrix (only works for 2D)
    y : array-like, shape (n_samples,)
        Target labels
    classifier : fitted classifier object
        Must have predict method
    title : str
        Plot title
    feature_names : list
        Names for x and y axes
    class_names : list
        Names for classes
    support_vectors : array-like
        Support vectors to highlight
    """
    if X.shape[1] != 2:
        raise ValueError("This function only works for 2D data")
    
    plt.figure(figsize=(10, 8))
    
    # Create a mesh for plotting decision boundary
    h = 0.02  # step size in the mesh
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    
    # Predict on mesh points
    mesh_points = np.c_[xx.ravel(), yy.ravel()]
    Z = classifier.predict(mesh_points)
    Z = Z.reshape(xx.shape)
    
    # Plot decision boundary
    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.RdYlBu)
    plt.contour(xx, yy, Z, colors='black', linestyles='--', linewidths=1)
    
    # Plot data points
    scatter = plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.RdYlBu, edgecolors='black')
    
    # Highlight support vectors if provided
    if support_vectors is not None:
        plt.scatter(support_vectors[:, 0], support_vectors[:, 1], 
                   s=300, linewidth=3, facecolors='none', edgecolors='red', 
                   label='Support Vectors')
        plt.legend()
    
    # Labels and title
    plt.xlabel(feature_names[0] if feature_names else 'Feature 1')
    plt.ylabel(feature_names[1] if feature_names else 'Feature 2')
    plt.title(title)
    
    # Add colorbar
    cbar = plt.colorbar(scatter)
    if class_names:
        cbar.set_label('Class')
        cbar.set_ticks([0, 1])
        cbar.set_ticklabels(class_names)
    
    plt.tight_layout()
    plt.show()

def plot_svm_margins(X, y, classifier, title="SVM with Margins"):
    """
    Plot SVM decision boundary with margins and support vectors
    """
    if X.shape[1] != 2:
        raise ValueError("This function only works for 2D data")
    
    plt.figure(figsize=(10, 8))
    
    # Create mesh
    h = 0.02
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    
    # Get decision function values
    mesh_points = np.c_[xx.ravel(), yy.ravel()]
    Z = classifier.decision_function(mesh_points)
    Z = Z.reshape(xx.shape)
    
    # Plot decision boundary and margins
    plt.contour(xx, yy, Z, colors='k', levels=[-1, 0, 1], 
               alpha=0.5, linestyles=['--', '-', '--'])
    
    # Fill margin area
    plt.contourf(xx, yy, Z, levels=[-1, 1], alpha=0.1, colors=['red'])
    
    # Plot data points
    colors = ['red', 'blue']
    for i, color in enumerate(colors):
        idx = np.where(y == i)
        plt.scatter(X[idx, 0], X[idx, 1], c=color, marker='o', s=50, 
                   edgecolors='black', label=f'Class {i}')
    
    # Highlight support vectors
    if hasattr(classifier, 'support_vectors_'):
        plt.scatter(classifier.support_vectors_[:, 0], 
                   classifier.support_vectors_[:, 1],
                   s=300, linewidth=3, facecolors='none', 
                   edgecolors='orange', label='Support Vectors')
    
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title(title)
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

def generate_sample_data(n_samples=100, noise=0.1, random_state=42):
    """Generate sample datasets for SVM examples"""
    np.random.seed(random_state)
    
    # Linear separable data
    linear_X, linear_y = make_classification(n_samples=n_samples, n_features=2, 
                                           n_redundant=0, n_informative=2,
                                           n_clusters_per_class=1, 
                                           random_state=random_state)
    
    # Non-linear data (circles)
    theta = np.linspace(0, 2*np.pi, n_samples//2)
    r1, r2 = 1, 3
    circle_X1 = np.column_stack([r1*np.cos(theta), r1*np.sin(theta)])
    circle_X2 = np.column_stack([r2*np.cos(theta), r2*np.sin(theta)])
    circle_X = np.vstack([circle_X1, circle_X2])
    circle_y = np.hstack([np.zeros(n_samples//2), np.ones(n_samples//2)])
    
    # Add noise
    circle_X += np.random.normal(0, noise, circle_X.shape)
    
    return linear_X, linear_y, circle_X, circle_y

def evaluate_classifier(classifier, X_test, y_test, class_names=None):
    """Comprehensive evaluation of a classifier"""
    y_pred = classifier.predict(X_test)
    
    print("üìä Classification Results:")
    print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
    print(f"Precision: {precision_score(y_test, y_pred, average='weighted'):.4f}")
    print(f"Recall: {recall_score(y_test, y_pred, average='weighted'):.4f}")
    print(f"F1-Score: {f1_score(y_test, y_pred, average='weighted'):.4f}")
    
    print("\nüìã Detailed Classification Report:")
    print(classification_report(y_test, y_pred, target_names=class_names))
    
    return y_pred

print("üîß Utility functions loaded successfully!")
````

## üìê Vectors and Vector Spaces

### What is a Vector?

A **vector** is a mathematical object that has both magnitude and direction. In the context of machine learning and SVM, vectors represent:

- **Feature vectors**: Data points in our dataset
- **Weight vectors**: Parameters of our model
- **Normal vectors**: Directions perpendicular to hyperplanes

### Vector Operations

Let's explore the fundamental vector operations with practical examples:


````python
# Vector operations demonstration
print("üî¢ Vector Operations for SVM")
print("=" * 50)

# Define sample vectors
v1 = np.array([3, 4])
v2 = np.array([1, 2])
v3 = np.array([2, -1])

print(f"Vector v1: {v1}")
print(f"Vector v2: {v2}")
print(f"Vector v3: {v3}")
print()

# 1. Vector Addition
print("1. Vector Addition:")
v_add = v1 + v2
print(f"v1 + v2 = {v_add}")
print()

# 2. Scalar Multiplication
print("2. Scalar Multiplication:")
scalar = 2
v_scaled = scalar * v1
print(f"{scalar} * v1 = {v_scaled}")
print()

# 3. Vector Magnitude (L2 Norm)
print("3. Vector Magnitude (L2 Norm):")
magnitude_v1 = np.linalg.norm(v1)
magnitude_v2 = np.linalg.norm(v2)
print(f"||v1|| = {magnitude_v1:.3f}")
print(f"||v2|| = {magnitude_v2:.3f}")
print()

# 4. Unit Vector (Normalization)
print("4. Unit Vector (Normalization):")
v1_unit = v1 / magnitude_v1
print(f"v1 normalized: {v1_unit}")
print(f"Magnitude of normalized v1: {np.linalg.norm(v1_unit):.3f}")
print()

# 5. Dot Product - CRUCIAL for SVM!
print("5. Dot Product (CRUCIAL for SVM!):")
dot_product = np.dot(v1, v2)
print(f"v1 ¬∑ v2 = {dot_product}")
print()

# Geometric interpretation of dot product
angle_rad = np.arccos(dot_product / (magnitude_v1 * magnitude_v2))
angle_deg = np.degrees(angle_rad)
print(f"Angle between v1 and v2: {angle_deg:.2f} degrees")
print()

# 6. Visualization of vectors
plt.figure(figsize=(12, 5))

# Plot 1: Vector addition and scaling
plt.subplot(1, 2, 1)
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='v1', width=0.006)
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red', label='v2', width=0.006)
plt.quiver(0, 0, v_add[0], v_add[1], angles='xy', scale_units='xy', scale=1, color='green', label='v1+v2', width=0.006)
plt.quiver(v1[0], v1[1], v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red', alpha=0.5, width=0.003)
plt.grid(True, alpha=0.3)
plt.axis('equal')
plt.xlim(-1, 6)
plt.ylim(-1, 7)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Vector Addition')
plt.legend()

# Plot 2: Dot product visualization
plt.subplot(1, 2, 2)
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', label=f'v1', width=0.006)
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red', label=f'v2', width=0.006)

# Draw angle arc
theta = np.linspace(0, angle_rad, 50)
arc_x = 0.5 * np.cos(theta)
arc_y = 0.5 * np.sin(theta)
plt.plot(arc_x, arc_y, 'k--', alpha=0.5)
plt.text(0.7, 0.3, f'{angle_deg:.1f}¬∞', fontsize=10)

plt.grid(True, alpha=0.3)
plt.axis('equal')
plt.xlim(-1, 4)
plt.ylim(-1, 5)
plt.xlabel('X')
plt.ylabel('Y')
plt.title(f'Dot Product = {dot_product}')
plt.legend()

plt.tight_layout()
plt.show()

print("üí° Key Insights:")
print("‚Ä¢ Dot product measures how much two vectors point in the same direction")
print("‚Ä¢ When dot product = 0, vectors are perpendicular (orthogonal)")
print("‚Ä¢ When dot product > 0, vectors point in similar directions")
print("‚Ä¢ When dot product < 0, vectors point in opposite directions")
print("‚Ä¢ This concept is fundamental to understanding SVM hyperplanes!")
````

````python
# Extended Dot Product Examples for SVM Understanding
print("üî¢ Extended Dot Product Examples for SVM")
print("=" * 50)

# Example 1: Orthogonal vectors (perpendicular)
print("1. Orthogonal Vectors (Perpendicular)")
print("-" * 40)
v_ortho1 = np.array([1, 0])
v_ortho2 = np.array([0, 1])
dot_ortho = np.dot(v_ortho1, v_ortho2)

print(f"v1 = {v_ortho1}, v2 = {v_ortho2}")
print(f"Dot product: {dot_ortho}")
print(f"Angle: {np.degrees(np.arccos(dot_ortho / (np.linalg.norm(v_ortho1) * np.linalg.norm(v_ortho2)))):.1f}¬∞")
print("üí° When dot product = 0, vectors are perpendicular!")
print()

# Example 2: Parallel vectors (same direction)
print("2. Parallel Vectors (Same Direction)")
print("-" * 35)
v_parallel1 = np.array([2, 3])
v_parallel2 = np.array([4, 6])  # 2 * v_parallel1
dot_parallel = np.dot(v_parallel1, v_parallel2)

print(f"v1 = {v_parallel1}, v2 = {v_parallel2}")
print(f"Dot product: {dot_parallel}")
angle_parallel = np.arccos(dot_parallel / (np.linalg.norm(v_parallel1) * np.linalg.norm(v_parallel2)))
print(f"Angle: {np.degrees(angle_parallel):.1f}¬∞")
print("üí° When vectors are parallel, angle = 0¬∞ and dot product is maximized!")
print()

# Example 3: Opposite direction vectors
print("3. Opposite Direction Vectors")
print("-" * 30)
v_opposite1 = np.array([1, 2])
v_opposite2 = np.array([-2, -4])  # -2 * v_opposite1
dot_opposite = np.dot(v_opposite1, v_opposite2)

print(f"v1 = {v_opposite1}, v2 = {v_opposite2}")
print(f"Dot product: {dot_opposite}")
if dot_opposite < 0:
    angle_opposite = np.arccos(dot_opposite / (np.linalg.norm(v_opposite1) * np.linalg.norm(v_opposite2)))
    print(f"Angle: {np.degrees(angle_opposite):.1f}¬∞")
print("üí° When dot product < 0, vectors point in opposite directions!")
print()

# Example 4: Data points and hyperplane normal (SVM context)
print("4. SVM Context: Data Points and Hyperplane Normal")
print("-" * 50)

# Define a hyperplane normal vector
w_normal = np.array([1, -1])  # Normal to hyperplane x + (-1)*y = 0, i.e., y = x
print(f"Hyperplane normal vector w: {w_normal}")
print(f"Hyperplane equation: {w_normal[0]}*x + {w_normal[1]}*y = 0")
print()

# Test points on different sides of hyperplane
test_points = np.array([
    [2, 1],    # Above the line y = x
    [1, 2],    # Below the line y = x  
    [1, 1],    # On the line y = x
    [3, 0],    # Above the line y = x
    [0, 3]     # Below the line y = x
])

point_names = ['Point A [2,1]', 'Point B [1,2]', 'Point C [1,1]', 'Point D [3,0]', 'Point E [0,3]']

for i, (point, name) in enumerate(zip(test_points, point_names)):
    dot_val = np.dot(w_normal, point)
    side = "positive side" if dot_val > 0 else "negative side" if dot_val < 0 else "on hyperplane"
    print(f"{name}: w¬∑x = {w_normal} ¬∑ {point} = {dot_val} ‚Üí {side}")

print()

# Example 5: Feature similarity using dot product
print("5. Feature Similarity in Machine Learning")
print("-" * 40)

# Simulated feature vectors for different samples
sample1 = np.array([0.8, 0.6, 0.2])  # High values in first two features
sample2 = np.array([0.7, 0.5, 0.1])  # Similar pattern to sample1
sample3 = np.array([0.1, 0.2, 0.9])  # Different pattern

print(f"Sample 1: {sample1}")
print(f"Sample 2: {sample2}")
print(f"Sample 3: {sample3}")
print()

# Calculate pairwise similarities
sim_1_2 = np.dot(sample1, sample2)
sim_1_3 = np.dot(sample1, sample3)
sim_2_3 = np.dot(sample2, sample3)

print(f"Similarity (sample1, sample2): {sim_1_2:.3f}")
print(f"Similarity (sample1, sample3): {sim_1_3:.3f}")
print(f"Similarity (sample2, sample3): {sim_2_3:.3f}")
print("üí° Higher dot product = more similar feature patterns!")
print()

# Example 6: Geometric interpretation with visualization
print("6. Geometric Interpretation Visualization")
print("-" * 40)

# Create comprehensive visualization
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# Plot 1: Orthogonal vectors
ax = axes[0, 0]
ax.quiver(0, 0, v_ortho1[0], v_ortho1[1], angles='xy', scale_units='xy', scale=1, 
          color='blue', width=0.01, label=f'v1={v_ortho1}')
ax.quiver(0, 0, v_ortho2[0], v_ortho2[1], angles='xy', scale_units='xy', scale=1, 
          color='red', width=0.01, label=f'v2={v_ortho2}')
ax.set_xlim(-0.5, 1.5)
ax.set_ylim(-0.5, 1.5)
ax.grid(True, alpha=0.3)
ax.set_title(f'Orthogonal Vectors\nDot Product = {dot_ortho}')
ax.legend()
ax.set_aspect('equal')

# Plot 2: Parallel vectors
ax = axes[0, 1]
ax.quiver(0, 0, v_parallel1[0], v_parallel1[1], angles='xy', scale_units='xy', scale=1, 
          color='blue', width=0.008, label=f'v1={v_parallel1}')
ax.quiver(0, 0, v_parallel2[0], v_parallel2[1], angles='xy', scale_units='xy', scale=1, 
          color='red', width=0.008, label=f'v2={v_parallel2}')
ax.set_xlim(-1, 5)
ax.set_ylim(-1, 7)
ax.grid(True, alpha=0.3)
ax.set_title(f'Parallel Vectors\nDot Product = {dot_parallel}')
ax.legend()
ax.set_aspect('equal')

# Plot 3: Opposite direction vectors
ax = axes[0, 2]
ax.quiver(0, 0, v_opposite1[0], v_opposite1[1], angles='xy', scale_units='xy', scale=1, 
          color='blue', width=0.01, label=f'v1={v_opposite1}')
ax.quiver(0, 0, v_opposite2[0], v_opposite2[1], angles='xy', scale_units='xy', scale=1, 
          color='red', width=0.01, label=f'v2={v_opposite2}')
ax.set_xlim(-3, 2)
ax.set_ylim(-5, 3)
ax.grid(True, alpha=0.3)
ax.set_title(f'Opposite Vectors\nDot Product = {dot_opposite}')
ax.legend()
ax.set_aspect('equal')

# Plot 4: SVM hyperplane context
ax = axes[1, 0]
# Plot hyperplane y = x
x_line = np.linspace(-1, 4, 100)
y_line = x_line
ax.plot(x_line, y_line, 'k-', linewidth=2, label='Hyperplane: y = x')

# Plot normal vector
ax.quiver(1.5, 1.5, w_normal[0], w_normal[1], angles='xy', scale_units='xy', scale=1, 
          color='purple', width=0.01, label=f'Normal w={w_normal}')

# Plot test points
colors = ['red' if np.dot(w_normal, point) > 0 else 'blue' for point in test_points]
ax.scatter(test_points[:, 0], test_points[:, 1], c=colors, s=100, edgecolors='black')

# Annotate points
for i, (point, name) in enumerate(zip(test_points, point_names)):
    ax.annotate(f'{name[6:]}', (point[0], point[1]), xytext=(5, 5), 
                textcoords='offset points', fontsize=10)

ax.set_xlim(-0.5, 3.5)
ax.set_ylim(-0.5, 3.5)
ax.grid(True, alpha=0.3)
ax.set_title('SVM Hyperplane Context')
ax.legend()
ax.set_aspect('equal')

# Plot 5: Angle vs dot product relationship
ax = axes[1, 1]
angles = np.linspace(0, np.pi, 100)
dot_products = np.cos(angles)  # For unit vectors

ax.plot(np.degrees(angles), dot_products, 'b-', linewidth=2)
ax.set_xlabel('Angle (degrees)')
ax.set_ylabel('Dot Product (unit vectors)')
ax.set_title('Angle vs Dot Product Relationship')
ax.grid(True, alpha=0.3)
ax.axhline(y=0, color='red', linestyle='--', alpha=0.7, label='Orthogonal')
ax.axvline(x=90, color='red', linestyle='--', alpha=0.7)
ax.legend()

# Plot 6: Feature similarity heatmap
ax = axes[1, 2]
samples = np.array([sample1, sample2, sample3])
similarity_matrix = samples @ samples.T

im = ax.imshow(similarity_matrix, cmap='viridis', aspect='equal')
ax.set_title('Feature Similarity Matrix\n(Dot Products)')
ax.set_xticks([0, 1, 2])
ax.set_yticks([0, 1, 2])
ax.set_xticklabels(['Sample 1', 'Sample 2', 'Sample 3'])
ax.set_yticklabels(['Sample 1', 'Sample 2', 'Sample 3'])

# Add text annotations
for i in range(3):
    for j in range(3):
        text = ax.text(j, i, f'{similarity_matrix[i, j]:.2f}',
                      ha="center", va="center", color="white", fontweight='bold')

plt.colorbar(im, ax=ax)
plt.tight_layout()
plt.show()

# Example 7: Practical SVM decision making
print("7. Practical SVM Decision Making with Dot Products")
print("-" * 50)

# Simulate a trained SVM scenario
w_svm = np.array([2, -1])  # Learned weight vector
b_svm = 0.5                # Learned bias

print(f"Trained SVM parameters:")
print(f"Weight vector w: {w_svm}")
print(f"Bias b: {b_svm}")
print(f"Decision function: f(x) = w¬∑x + b = {w_svm[0]}*x‚ÇÅ + {w_svm[1]}*x‚ÇÇ + {b_svm}")
print()

# Test new data points
new_points = np.array([
    [1, 1],
    [0, 2], 
    [2, 0],
    [-1, 0],
    [1, 3]
])

print("Classifying new data points:")
for i, point in enumerate(new_points):
    dot_product = np.dot(w_svm, point)
    decision_value = dot_product + b_svm
    prediction = "Class +1" if decision_value > 0 else "Class -1"
    
    print(f"Point {point}:")
    print(f"  w¬∑x = {w_svm}¬∑{point} = {dot_product}")
    print(f"  f(x) = {dot_product} + {b_svm} = {decision_value}")
    print(f"  Prediction: {prediction}")
    print()

print("üîë Key Insights:")
print("‚Ä¢ Dot product measures alignment between vectors")
print("‚Ä¢ In SVM: w¬∑x tells us which side of hyperplane point x is on")
print("‚Ä¢ Larger |w¬∑x| means point is farther from decision boundary")
print("‚Ä¢ Sign of w¬∑x + b determines final classification")
print("‚Ä¢ Understanding dot products is crucial for SVM intuition!")
````

## üéØ Hyperplanes: The Heart of SVM

A **hyperplane** is a fundamental concept in SVM. It's a generalization of a line (in 2D) or a plane (in 3D) to higher dimensions.

### Mathematical Definition

In an n-dimensional space, a hyperplane is defined by the equation:

$$w_1x_1 + w_2x_2 + ... + w_nx_n + b = 0$$

Or in vector notation:

$$\mathbf{w}^T\mathbf{x} + b = 0$$

Where:

- $\mathbf{w} = [w_1, w_2, ..., w_n]$ is the **weight vector** (normal to the hyperplane)
- $\mathbf{x} = [x_1, x_2, ..., x_n]$ is a **point** on the hyperplane
- $b$ is the **bias term** (determines the offset from origin)

### Key Properties:

1. **Normal Vector**: $\mathbf{w}$ is perpendicular to the hyperplane
2. **Distance from Origin**: $|b|/||\mathbf{w}||$
3. **Decision Function**: Points on different sides have different signs for $\mathbf{w}^T\mathbf{x} + b$

Let's visualize hyperplanes in 2D and 3D:


````python
# Hyperplane visualization and analysis
print("üéØ Hyperplane Analysis")
print("=" * 40)

# Define a hyperplane in 2D: w1*x1 + w2*x2 + b = 0
w = np.array([1, 2])  # Normal vector
b = -3               # Bias term

print(f"Hyperplane equation: {w[0]}*x1 + {w[1]}*x2 + {b} = 0")
print(f"Normal vector w: {w}")
print(f"Bias term b: {b}")
print()

# Calculate some properties
w_magnitude = np.linalg.norm(w)
distance_from_origin = abs(b) / w_magnitude
w_unit = w / w_magnitude

print(f"||w||: {w_magnitude:.3f}")
print(f"Distance from origin: {distance_from_origin:.3f}")
print(f"Unit normal vector: {w_unit}")
print()

# Generate points for plotting the line
x1_range = np.linspace(-5, 5, 100)
# From w1*x1 + w2*x2 + b = 0, solve for x2
x2_line = -(w[0] * x1_range + b) / w[1]

# Generate test points to show decision function
np.random.seed(42)
test_points = np.random.uniform(-4, 4, (20, 2))
decision_values = test_points @ w + b

# Visualization
plt.figure(figsize=(15, 5))

# Plot 1: 2D Hyperplane (line) with normal vector
plt.subplot(1, 3, 1)
plt.plot(x1_range, x2_line, 'b-', linewidth=2, label='Hyperplane')

# Plot normal vector from origin
plt.arrow(0, 0, w[0], w[1], head_width=0.2, head_length=0.2, fc='red', ec='red', label='Normal vector w')

# Plot normal vector from a point on the line
point_on_line = np.array([0, -b/w[1]])
plt.arrow(point_on_line[0], point_on_line[1], w_unit[0], w_unit[1], 
          head_width=0.15, head_length=0.15, fc='orange', ec='orange', 
          label='Unit normal')

plt.scatter(*point_on_line, color='green', s=100, label='Point on hyperplane')
plt.grid(True, alpha=0.3)
plt.axis('equal')
plt.xlim(-5, 5)
plt.ylim(-3, 3)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('2D Hyperplane with Normal Vector')
plt.legend()

# Plot 2: Decision function visualization
plt.subplot(1, 3, 2)
positive_points = test_points[decision_values > 0]
negative_points = test_points[decision_values <= 0]

plt.scatter(positive_points[:, 0], positive_points[:, 1], c='red', marker='o', 
           s=50, label='f(x) > 0', alpha=0.7)
plt.scatter(negative_points[:, 0], negative_points[:, 1], c='blue', marker='s', 
           s=50, label='f(x) ‚â§ 0', alpha=0.7)
plt.plot(x1_range, x2_line, 'k-', linewidth=2, label='Decision boundary')

plt.grid(True, alpha=0.3)
plt.xlim(-5, 5)
plt.ylim(-4, 4)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Decision Function: w^T x + b')
plt.legend()

# Plot 3: Distance from points to hyperplane
plt.subplot(1, 3, 3)
# Calculate distances from test points to hyperplane
distances = np.abs(decision_values) / w_magnitude

plt.scatter(test_points[:, 0], test_points[:, 1], c=distances, 
           cmap='viridis', s=100, alpha=0.7)
plt.colorbar(label='Distance to hyperplane')
plt.plot(x1_range, x2_line, 'k-', linewidth=2, label='Hyperplane')

# Show some distance lines
for i in range(0, len(test_points), 3):
    point = test_points[i]
    # Project point onto hyperplane
    projection = point - (decision_values[i] / (w_magnitude**2)) * w
    plt.plot([point[0], projection[0]], [point[1], projection[1]], 
             'r--', alpha=0.5, linewidth=1)

plt.grid(True, alpha=0.3)
plt.xlim(-5, 5)
plt.ylim(-4, 4)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Distances to Hyperplane')

plt.tight_layout()
plt.show()

# Mathematical formulations
print("üìê Distance Formula:")
print("Distance from point x to hyperplane: |w^T x + b| / ||w||")
print()

print("üéØ Decision Function:")
print("f(x) = w^T x + b")
print("‚Ä¢ f(x) > 0: Point is on positive side")
print("‚Ä¢ f(x) < 0: Point is on negative side")  
print("‚Ä¢ f(x) = 0: Point is on the hyperplane")
print()

print("Example calculations:")
sample_point = np.array([1, 1])
decision_val = sample_point @ w + b
distance_val = abs(decision_val) / w_magnitude
print(f"For point {sample_point}:")
print(f"  Decision function: f({sample_point}) = {decision_val}")
print(f"  Distance to hyperplane: {distance_val:.3f}")
````

## üìè Mathematical Notation: Single vs Double Bars

### Important Distinction in Mathematical Notation

In mathematics and machine learning, the notation with bars has specific meanings that are crucial to understand:

---

### **Single Bars: | |**

**Absolute Value** (for scalars) or **Determinant** (for matrices)

- **For scalars**: $|x|$ = absolute value of $x$

  - Example: $|-5| = 5$, $|3.7| = 3.7$

- **For matrices**: $|A|$ = determinant of matrix $A$
  - Example: $\begin{vmatrix} 2 & 1 \\ 3 & 4 \end{vmatrix} = 8 - 3 = 5$

---

### **Double Bars: || ||**

**Norm** (magnitude/length) of a vector or matrix

#### **Vector Norms:**

- **L2 Norm (Euclidean)**: $||\mathbf{v}||_2 = ||\mathbf{v}|| = \sqrt{v_1^2 + v_2^2 + ... + v_n^2}$

  - Most common, often written without subscript
  - Example: $||\begin{bmatrix} 3 \\ 4 \end{bmatrix}|| = \sqrt{3^2 + 4^2} = 5$

- **L1 Norm (Manhattan)**: $||\mathbf{v}||_1 = |v_1| + |v_2| + ... + |v_n|$
  - Example: $||\begin{bmatrix} 3 \\ 4 \end{bmatrix}||_1 = |3| + |4| = 7$

#### **In SVM Context:**

- $||\mathbf{w}||$ = magnitude of weight vector
- $||\mathbf{w}||^2 = \mathbf{w}^T\mathbf{w}$ = squared magnitude
- Margin = $\frac{2}{||\mathbf{w}||}$

---

### **Key Examples in This Course:**

```python
# Single bars - absolute value
distance_to_hyperplane = |w^T x + b| / ||w||

# Double bars - vector norm
margin = 2 / ||w||
objective_function = (1/2) * ||w||¬≤
```

---

### **Memory Aid:**

- **Single bars** `| |` ‚Üí **Absolute value** (scalar concept)
- **Double bars** `|| ||` ‚Üí **Norm/Length** (vector concept)

This distinction is fundamental in SVM mathematics, where we frequently work with both absolute values of scalar expressions and norms of weight vectors!

## üîß Optimization Problems

### Types of Optimization Problems

1. **Unconstrained Optimization**:

   - Minimize $f(x)$ without restrictions
   - Example: Linear regression with least squares

2. **Constrained Optimization**:
   - Minimize $f(x)$ subject to constraints $g_i(x) \leq 0$ and $h_j(x) = 0$
   - Example: SVM margin maximization with classification constraints

### The SVM Optimization Problem Preview

SVM solves the following optimization problem:

$$\min_{w,b} \frac{1}{2}||w||^2$$

Subject to: $y_i(w^T x_i + b) \geq 1$ for all $i$

This is a **quadratic programming** problem with **linear constraints**.

Let's explore optimization concepts with hands-on examples:


## üîÑ Recap: Lagrange Multipliers for SVM

### üìö What are Lagrange Multipliers?

Lagrange multipliers are a mathematical technique for solving **constrained optimization problems**. They transform a constrained problem into an unconstrained one by introducing auxiliary variables (the multipliers).

---

### üéØ The General Problem

**Standard Form:**

```
Minimize: f(x)
Subject to: g(x) = 0 (equality constraints)
            h(x) ‚â§ 0 (inequality constraints)
```

**Lagrangian Function:**
$$L(x, \lambda, \mu) = f(x) + \lambda g(x) + \mu h(x)$$

Where:

- $\lambda$ = Lagrange multipliers for equality constraints
- $\mu$ = Lagrange multipliers for inequality constraints

---

### üîß SVM Application

In SVM, we solve:

**Primal Problem:**

```
Minimize: (1/2)||w||¬≤
Subject to: yi(w^T xi + b) ‚â• 1  for all i
```

**Lagrangian for SVM:**
$$L(w, b, \alpha) = \frac{1}{2}||w||^2 - \sum_{i=1}^n \alpha_i [y_i(w^T x_i + b) - 1]$$

Where $\alpha_i \geq 0$ are the Lagrange multipliers.

---

### ‚ö° Key Optimality Conditions (KKT)

1. **Stationarity:** $\nabla L = 0$

   - $\frac{\partial L}{\partial w} = w - \sum_i \alpha_i y_i x_i = 0$
   - $\frac{\partial L}{\partial b} = -\sum_i \alpha_i y_i = 0$

2. **Primal Feasibility:** $y_i(w^T x_i + b) \geq 1$

3. **Dual Feasibility:** $\alpha_i \geq 0$

4. **Complementary Slackness:** $\alpha_i [y_i(w^T x_i + b) - 1] = 0$

---

### üí° Physical Interpretation in SVM

- **$\alpha_i = 0$:** Point is not a support vector (constraint inactive)
- **$\alpha_i > 0$:** Point is a support vector (constraint active)
- **Support vectors:** Points that lie exactly on the margin boundaries
- **Weight vector:** $w = \sum_i \alpha_i y_i x_i$ (linear combination of support vectors)

---

### üéâ Why This Matters for SVM

1. **Sparsity:** Most $\alpha_i = 0$, only support vectors matter
2. **Dual Problem:** Easier to solve than primal for high dimensions
3. **Kernel Trick:** Enables non-linear classification
4. **Memory Efficiency:** Only need to store support vectors

Lagrange multipliers are the mathematical foundation that makes SVM's "maximum margin" principle computationally tractable!


## üìê Recap: Quadratic Programming for SVM

### üéØ What is Quadratic Programming?

**Quadratic Programming (QP)** is a mathematical optimization technique where we:

- **Minimize** a quadratic objective function
- **Subject to** linear constraints

## This is exactly what SVM does!

### üßÆ General QP Form

**Standard Formulation:**

```
Minimize: (1/2) x^T P x + q^T x\
Subject to: G x ‚â§ h  (inequality constraints)
           A x = b   (equality constraints)
```

Where:

- **x** = variables to optimize
- **P** = positive semidefinite matrix (defines quadratic terms)
- **q** = linear coefficients vector
- **G, h** = inequality constraint matrices/vectors
- **A, b** = equality constraint matrices/vectors

---

### üéØ SVM as Quadratic Programming

#### **Primal SVM Problem:**

```
Minimize: (1/2)||w||¬≤
Subject to: yi(w^T xi + b) ‚â• 1  for all i
```

#### **Dual SVM Problem (QP Form):**

```
Minimize: (1/2) Œ±^T P Œ± - 1^T Œ±
Subject to: y^T Œ± = 0     (equality)
           Œ± ‚â• 0         (inequality)
```

Where:

- **Œ±** = Lagrange multipliers (variables to optimize)
- **P** = Kernel matrix K with P_ij = yi √ó yj √ó K(xi, xj)
- **1** = vector of ones
- **y** = label vector

---

### üîß Key Properties of QP

#### **1. Convex Optimization**

- Quadratic objective with linear constraints
- **Global minimum guaranteed** (no local minima)
- Efficient algorithms available

#### **2. Unique Solution**

- When P is positive definite ‚Üí unique solution
- SVM always finds the same optimal hyperplane

#### **3. Computational Efficiency**

- Well-studied problem class
- Specialized solvers (SMO, interior point methods)
- Polynomial time complexity

---

### üí° Why QP is Perfect for SVM

#### **1. Mathematical Properties**

- **Convexity** ensures global optimum
- **Duality theory** enables kernel trick
- **Sparse solutions** (most Œ±i = 0)

#### **2. Practical Advantages**

- **Numerical stability**
- **Efficient algorithms** (SMO, coordinate descent)
- **Handles constraints naturally**

#### **3. Theoretical Guarantees**

- **Optimality conditions** (KKT conditions)
- **Uniqueness** of maximum margin solution
- **Generalization bounds**

---

### üõ†Ô∏è QP Solving Methods for SVM

#### **1. Sequential Minimal Optimization (SMO)**

- Most popular for SVM
- Optimizes 2 variables at a time
- Memory efficient for large datasets

#### **2. Interior Point Methods**

- Good for medium-sized problems
- Guaranteed polynomial convergence
- Used in many QP solvers

#### **3. Coordinate Descent**

- Simple to implement
- Good for sparse problems
- Used in LIBSVM

---

### üìä QP vs Other Optimization Types

| Problem Type              | Objective | Constraints | Complexity  |
| ------------------------- | --------- | ----------- | ----------- |
| **Linear Programming**    | Linear    | Linear      | Polynomial  |
| **Quadratic Programming** | Quadratic | Linear      | Polynomial  |
| **Nonlinear Programming** | Nonlinear | Nonlinear   | Exponential |

**SVM uses QP** ‚Üí Best of both worlds: rich enough to capture margin maximization, simple enough for efficient solving.

---

### üîë Key Insights for SVM

1. **QP formulation** makes SVM optimization tractable
2. **Convexity** guarantees finding the global optimum
3. **Dual formulation** enables the kernel trick
4. **Sparsity** emerges naturally (support vectors)
5. **Efficient algorithms** make SVM scalable

**Bottom line:** Quadratic Programming is the mathematical foundation that makes SVM both theoretically elegant and practically efficient!


````python
# Optimization examples and visualization
from scipy.optimize import minimize
from scipy.optimize import minimize_scalar

print("üîß Optimization Theory for SVM")
print("=" * 50)

# Example 1: Simple unconstrained optimization
print("1. Unconstrained Optimization Example")
print("-" * 40)

def simple_quadratic(x):
    """Simple quadratic function: f(x) = (x-2)^2 + 1"""
    return (x - 2)**2 + 1

def quadratic_derivative(x):
    """Derivative: f'(x) = 2(x-2)"""
    return 2 * (x - 2)

# Find minimum using calculus (derivative = 0)
critical_point = 2  # Where f'(x) = 0
min_value = simple_quadratic(critical_point)

print(f"Function: f(x) = (x-2)¬≤ + 1")
print(f"Minimum at x = {critical_point}, f(x) = {min_value}")

# Numerical optimization
result = minimize_scalar(simple_quadratic)
print(f"Numerical solution: x = {result.x:.6f}, f(x) = {result.fun:.6f}")
print()

# Visualization
x_vals = np.linspace(-1, 5, 1000)
y_vals = [simple_quadratic(x) for x in x_vals]

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(x_vals, y_vals, 'b-', linewidth=2, label='f(x) = (x-2)¬≤ + 1')
plt.plot(critical_point, min_value, 'ro', markersize=10, label='Minimum')
plt.grid(True, alpha=0.3)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Unconstrained Optimization')
plt.legend()

# Example 2: Constrained optimization using Lagrange multipliers
print("2. Constrained Optimization with Lagrange Multipliers")
print("-" * 55)

def objective_2d(vars):
    """Objective function: f(x,y) = x¬≤ + y¬≤"""
    x, y = vars
    return x**2 + y**2

def constraint(vars):
    """Constraint: g(x,y) = x + y - 2 = 0"""
    x, y = vars
    return x + y - 2

# Using Lagrange multipliers analytically:
# L(x,y,Œª) = x¬≤ + y¬≤ - Œª(x + y - 2)
# ‚àÇL/‚àÇx = 2x - Œª = 0  ‚Üí  x = Œª/2
# ‚àÇL/‚àÇy = 2y - Œª = 0  ‚Üí  y = Œª/2
# ‚àÇL/‚àÇŒª = -(x + y - 2) = 0  ‚Üí  x + y = 2
# Substituting: Œª/2 + Œª/2 = 2  ‚Üí  Œª = 2
# Therefore: x = y = 1

analytical_solution = np.array([1, 1])
analytical_value = objective_2d(analytical_solution)

print("Problem: Minimize f(x,y) = x¬≤ + y¬≤ subject to x + y = 2")
print(f"Analytical solution: x = {analytical_solution[0]}, y = {analytical_solution[1]}")
print(f"Objective value: {analytical_value}")

# Numerical solution
constraints = {'type': 'eq', 'fun': constraint}
result_2d = minimize(objective_2d, [0, 0], constraints=constraints)
print(f"Numerical solution: x = {result_2d.x[0]:.6f}, y = {result_2d.x[1]:.6f}")
print(f"Objective value: {result_2d.fun:.6f}")
print()

# Visualization of constrained optimization
plt.subplot(1, 3, 2)

# Create grid for contour plot
x = np.linspace(-0.5, 2.5, 100)
y = np.linspace(-0.5, 2.5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

# Plot contours
contours = plt.contour(X, Y, Z, levels=20, alpha=0.6)
plt.clabel(contours, inline=True, fontsize=8)

# Plot constraint line
x_constraint = np.linspace(-0.5, 2.5, 100)
y_constraint = 2 - x_constraint
plt.plot(x_constraint, y_constraint, 'r-', linewidth=3, label='Constraint: x + y = 2')

# Plot solution
plt.plot(1, 1, 'ro', markersize=10, label='Optimal point (1,1)')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Constrained Optimization')
plt.legend()
plt.grid(True, alpha=0.3)
plt.axis('equal')

# Example 3: Quadratic Programming (SVM-like problem)
print("3. Quadratic Programming (SVM Preview)")
print("-" * 40)

# Simple QP problem: minimize (1/2)x^T Q x + c^T x
# Subject to: Ax ‚â§ b

# Define a simple 2D quadratic problem
Q = np.array([[2, 0], [0, 2]])  # Quadratic term matrix
c = np.array([-2, -2])          # Linear term
A = np.array([[1, 1], [-1, 0], [0, -1]])  # Inequality constraints matrix
b = np.array([1, 0, 0])         # Inequality bounds

def quadratic_objective(x):
    return 0.5 * x.T @ Q @ x + c.T @ x

def inequality_constraints(x):
    return A @ x - b

print("Quadratic Programming Problem:")
print("Minimize: (1/2)x^T Q x + c^T x")
print(f"Q = {Q}")
print(f"c = {c}")
print("Subject to: Ax ‚â§ b")
print(f"A = {A}")
print(f"b = {b}")
print()

# Solve using scipy
constraints_qp = {'type': 'ineq', 'fun': lambda x: -inequality_constraints(x)}
result_qp = minimize(quadratic_objective, [0, 0], constraints=constraints_qp)

print(f"Optimal solution: x = [{result_qp.x[0]:.3f}, {result_qp.x[1]:.3f}]")
print(f"Optimal value: {result_qp.fun:.3f}")

# Visualization of QP problem
plt.subplot(1, 3, 3)

# Create feasible region
x_qp = np.linspace(-0.5, 1.5, 100)
y_qp = np.linspace(-0.5, 1.5, 100)
X_qp, Y_qp = np.meshgrid(x_qp, y_qp)
Z_qp = 0.5 * (2*X_qp**2 + 2*Y_qp**2) + (-2*X_qp - 2*Y_qp)

# Plot objective function contours
contours_qp = plt.contour(X_qp, Y_qp, Z_qp, levels=15, alpha=0.6)

# Plot constraints
y1 = 1 - x_qp  # x + y ‚â§ 1
plt.fill_between(x_qp, np.maximum(0, np.maximum(y1, 0)), 1.5, alpha=0.2, color='gray', label='Feasible region')

# Plot constraint boundaries
plt.plot(x_qp, 1 - x_qp, 'g-', linewidth=2, label='x + y ‚â§ 1')
plt.axvline(x=0, color='g', linewidth=2)
plt.axhline(y=0, color='g', linewidth=2)

# Plot optimal solution
plt.plot(result_qp.x[0], result_qp.x[1], 'ro', markersize=10, label=f'Optimal ({result_qp.x[0]:.2f}, {result_qp.x[1]:.2f})')

plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Quadratic Programming')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nüí° Key Insights for SVM:")
print("‚Ä¢ SVM is fundamentally a quadratic programming problem")
print("‚Ä¢ We minimize ||w||¬≤ (quadratic) subject to linear constraints")
print("‚Ä¢ Lagrange multipliers transform constrained into unconstrained problem")
print("‚Ä¢ KKT conditions provide optimality criteria")
print("‚Ä¢ These concepts directly apply to SVM's dual formulation")
````

## üß† The Core Idea: Maximum Margin Classification

### The Problem Setup

Given a training dataset $\{(\mathbf{x}_i, y_i)\}_{i=1}^n$ where:

- $\mathbf{x}_i \in \mathbb{R}^d$ are feature vectors
- $y_i \in \{-1, +1\}$ are class labels

**Goal**: Find a hyperplane that separates the classes with maximum margin.

### Why Maximum Margin?

1. **Generalization**: Larger margins often lead to better performance on unseen data
2. **Uniqueness**: The maximum margin solution is unique
3. **Robustness**: Less sensitive to small perturbations in the data

### Mathematical Formulation

The decision boundary is defined by: $\mathbf{w}^T\mathbf{x} + b = 0$

For correct classification:

- If $y_i = +1$, then $\mathbf{w}^T\mathbf{x}_i + b > 0$
- If $y_i = -1$, then $\mathbf{w}^T\mathbf{x}_i + b < 0$

Combined constraint: $y_i(\mathbf{w}^T\mathbf{x}_i + b) > 0$ for all $i$

Let's visualize this concept:


````python
# Generate sample data for SVM illustration
np.random.seed(42)

# Create linearly separable data
n_samples = 20
X_class1 = np.random.multivariate_normal([2, 2], [[0.5, 0.1], [0.1, 0.5]], n_samples//2)
X_class2 = np.random.multivariate_normal([-1, -1], [[0.5, -0.1], [-0.1, 0.5]], n_samples//2)

X_demo = np.vstack([X_class1, X_class2])
y_demo = np.hstack([np.ones(n_samples//2), -np.ones(n_samples//2)])

print("üß† Maximum Margin Principle Visualization")
print("=" * 50)

# Multiple possible separating hyperplanes
hyperplanes = [
    {'w': np.array([1, 1]), 'b': 0, 'name': 'Hyperplane 1', 'color': 'red'},
    {'w': np.array([1, 0.5]), 'b': 0.5, 'name': 'Hyperplane 2', 'color': 'green'},
    {'w': np.array([0.8, 1.2]), 'b': -0.2, 'name': 'Hyperplane 3', 'color': 'blue'}
]

plt.figure(figsize=(15, 10))

# Plot 1: Multiple possible hyperplanes
plt.subplot(2, 2, 1)
colors = ['red', 'blue']
for i, color in enumerate(colors):
    mask = y_demo == (1 if i == 0 else -1)
    plt.scatter(X_demo[mask, 0], X_demo[mask, 1], c=color, marker='o', s=100, 
               label=f'Class {1 if i == 0 else -1}', edgecolors='black')

x_range = np.linspace(-4, 5, 100)
for hp in hyperplanes:
    x2_line = -(hp['w'][0] * x_range + hp['b']) / hp['w'][1]
    plt.plot(x_range, x2_line, color=hp['color'], linewidth=2, 
             linestyle='--', label=hp['name'])

plt.xlim(-4, 5)
plt.ylim(-4, 5)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Multiple Possible Separating Hyperplanes')
plt.legend()
plt.grid(True, alpha=0.3)

# Calculate margins for each hyperplane
plt.subplot(2, 2, 2)
margins = []

for j, hp in enumerate(hyperplanes):
    w, b = hp['w'], hp['b']
    w_norm = np.linalg.norm(w)
    
    # Calculate distances for each point
    distances = np.abs(X_demo @ w + b) / w_norm
    margin = np.min(distances)
    margins.append(margin)
    
    print(f"{hp['name']}:")
    print(f"  Normal vector: {w}")
    print(f"  Bias: {b}")
    print(f"  Margin: {margin:.3f}")
    print()

# Bar plot of margins
plt.bar(range(len(margins)), margins, color=[hp['color'] for hp in hyperplanes])
plt.xlabel('Hyperplane')
plt.ylabel('Margin')
plt.title('Margin Comparison')
plt.xticks(range(len(margins)), [hp['name'] for hp in hyperplanes])
plt.grid(True, alpha=0.3)

# Find the maximum margin hyperplane (simplified approach)
best_idx = np.argmax(margins)
best_hyperplane = hyperplanes[best_idx]
print(f"Best hyperplane: {best_hyperplane['name']} with margin {margins[best_idx]:.3f}")

# Plot 3: Margin visualization
plt.subplot(2, 2, 3)
colors = ['red', 'blue']
for i, color in enumerate(colors):
    mask = y_demo == (1 if i == 0 else -1)
    plt.scatter(X_demo[mask, 0], X_demo[mask, 1], c=color, marker='o', s=100, 
               label=f'Class {1 if i == 0 else -1}', edgecolors='black')

# Plot the best hyperplane and its margins
w_best, b_best = best_hyperplane['w'], best_hyperplane['b']
x_range = np.linspace(-4, 5, 100)

# Decision boundary
x2_boundary = -(w_best[0] * x_range + b_best) / w_best[1]
plt.plot(x_range, x2_boundary, 'k-', linewidth=3, label='Decision Boundary')

# Margin boundaries
margin_best = margins[best_idx]
w_norm_best = np.linalg.norm(w_best)

# Upper margin: w^T x + b = +1 (for +margin)
x2_upper = -(w_best[0] * x_range + b_best - margin_best * w_norm_best) / w_best[1]
# Lower margin: w^T x + b = -1 (for -margin)  
x2_lower = -(w_best[0] * x_range + b_best + margin_best * w_norm_best) / w_best[1]

plt.plot(x_range, x2_upper, 'k--', linewidth=2, alpha=0.7, label='Margin Boundaries')
plt.plot(x_range, x2_lower, 'k--', linewidth=2, alpha=0.7)

# Fill margin area
plt.fill_between(x_range, x2_lower, x2_upper, alpha=0.1, color='gray', label='Margin')

plt.xlim(-4, 5)
plt.ylim(-4, 5)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title(f'Maximum Margin Hyperplane (Margin = {margin_best:.3f})')
plt.legend()
plt.grid(True, alpha=0.3)

# Plot 4: Support vectors identification
plt.subplot(2, 2, 4)
colors = ['red', 'blue']

# Find approximate support vectors (points closest to boundary)
distances_to_boundary = np.abs(X_demo @ w_best + b_best) / w_norm_best
tolerance = 0.1  # tolerance for identifying support vectors
support_vector_mask = distances_to_boundary <= (margin_best + tolerance)

for i, color in enumerate(colors):
    class_mask = y_demo == (1 if i == 0 else -1)
    sv_mask = support_vector_mask & class_mask
    regular_mask = class_mask & ~support_vector_mask
    
    # Plot regular points
    if np.any(regular_mask):
        plt.scatter(X_demo[regular_mask, 0], X_demo[regular_mask, 1], 
                   c=color, marker='o', s=100, alpha=0.7, edgecolors='black')
    
    # Plot support vectors
    if np.any(sv_mask):
        plt.scatter(X_demo[sv_mask, 0], X_demo[sv_mask, 1], 
                   c=color, marker='o', s=200, edgecolors='yellow', linewidth=3,
                   label=f'Support Vectors (Class {1 if i == 0 else -1})')

# Plot boundaries
plt.plot(x_range, x2_boundary, 'k-', linewidth=3, label='Decision Boundary')
plt.plot(x_range, x2_upper, 'k--', linewidth=2, alpha=0.7)
plt.plot(x_range, x2_lower, 'k--', linewidth=2, alpha=0.7)

plt.xlim(-4, 5)
plt.ylim(-4, 5)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Support Vector Identification')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\nüìç Support Vectors Found: {np.sum(support_vector_mask)} out of {len(X_demo)} points")
print("\nüí° Key Observations:")
print("‚Ä¢ Only points closest to the decision boundary matter (support vectors)")
print("‚Ä¢ Maximum margin provides the most robust separation")
print("‚Ä¢ Support vectors fully determine the hyperplane")
print("‚Ä¢ Interior points can be removed without changing the solution")
````

## üìê Mathematical Derivation of Linear SVM

### Step 1: Canonical Hyperplane

To make the optimization problem well-defined, we use the **canonical form** where the closest points to the hyperplane satisfy:

$$y_i(\mathbf{w}^T\mathbf{x}_i + b) = 1$$

This means:

- Points on the positive margin: $\mathbf{w}^T\mathbf{x} + b = +1$
- Points on the negative margin: $\mathbf{w}^T\mathbf{x} + b = -1$
- All other points satisfy: $y_i(\mathbf{w}^T\mathbf{x}_i + b) \geq 1$

### Step 2: Margin Calculation

The **margin** is the distance between the two margin hyperplanes:

$$\text{margin} = \frac{2}{||\mathbf{w}||}$$

**Proof**:

- Distance from origin to $\mathbf{w}^T\mathbf{x} + b = +1$ is $\frac{|1-b|}{||\mathbf{w}||}$
- Distance from origin to $\mathbf{w}^T\mathbf{x} + b = -1$ is $\frac{|-1-b|}{||\mathbf{w}||}$
- Distance between parallel hyperplanes is $\frac{2}{||\mathbf{w}||}$

### Step 3: Primal Optimization Problem

**Maximize margin** ‚â° **Minimize** $\frac{1}{2}||\mathbf{w}||^2$

$$
\begin{align}
\min_{\mathbf{w},b} &\quad \frac{1}{2}||\mathbf{w}||^2 \\
\text{subject to} &\quad y_i(\mathbf{w}^T\mathbf{x}_i + b) \geq 1, \quad i = 1, ..., n
\end{align}
$$

This is a **quadratic programming** problem with **linear constraints**.

### Step 4: Lagrangian Formulation

Introduce Lagrange multipliers $\alpha_i \geq 0$:

$$L(\mathbf{w}, b, \boldsymbol{\alpha}) = \frac{1}{2}||\mathbf{w}||^2 - \sum_{i=1}^n \alpha_i [y_i(\mathbf{w}^T\mathbf{x}_i + b) - 1]$$

### Step 5: KKT Conditions

The Karush-Kuhn-Tucker (KKT) conditions for optimality:

1. **Stationarity**:
   $$\frac{\partial L}{\partial \mathbf{w}} = \mathbf{w} - \sum_{i=1}^n \alpha_i y_i \mathbf{x}_i = 0 \Rightarrow \mathbf{w} = \sum_{i=1}^n \alpha_i y_i \mathbf{x}_i$$
   $$\frac{\partial L}{\partial b} = -\sum_{i=1}^n \alpha_i y_i = 0 \Rightarrow \sum_{i=1}^n \alpha_i y_i = 0$$

2. **Primal feasibility**: $y_i(\mathbf{w}^T\mathbf{x}_i + b) \geq 1$

3. **Dual feasibility**: $\alpha_i \geq 0$

4. **Complementary slackness**: $\alpha_i [y_i(\mathbf{w}^T\mathbf{x}_i + b) - 1] = 0$

Let's implement these concepts step by step:


````python
# Mathematical derivation implementation
print("üìê SVM Mathematical Derivation - Step by Step")
print("=" * 55)

# Simple 2D example for clear illustration
X_simple = np.array([[1, 1], [2, 2], [2, 1], [-1, -1], [-2, -1], [-1, -2]])
y_simple = np.array([1, 1, 1, -1, -1, -1])

print("Dataset:")
for i in range(len(X_simple)):
    print(f"x_{i+1} = {X_simple[i]}, y_{i+1} = {y_simple[i]:2d}")
print()

# Manually solve for simple case to illustrate concepts
print("1. Finding the Optimal Hyperplane")
print("-" * 35)

# For this simple example, we'll use scikit-learn to find the solution
# and then verify our mathematical understanding
from sklearn.svm import SVC

svm_solver = SVC(kernel='linear', C=1e10)  # Very large C for hard margin
svm_solver.fit(X_simple, y_simple)

w_optimal = svm_solver.coef_[0]
b_optimal = svm_solver.intercept_[0]

print(f"Optimal weight vector w: [{w_optimal[0]:.3f}, {w_optimal[1]:.3f}]")
print(f"Optimal bias b: {b_optimal:.3f}")
print(f"||w||: {np.linalg.norm(w_optimal):.3f}")
print(f"Margin: {2/np.linalg.norm(w_optimal):.3f}")
print()

# Verify KKT conditions
print("2. Verifying KKT Conditions")
print("-" * 28)

# Get support vectors and alpha values
support_vectors = svm_solver.support_vectors_
support_indices = svm_solver.support_
alpha_sv = svm_solver.dual_coef_[0]  # Alpha values for support vectors
n_sv = len(support_vectors)

print(f"Number of support vectors: {n_sv}")
print("Support vectors:")
for i, (sv, alpha) in enumerate(zip(support_vectors, alpha_sv)):
    sv_idx = support_indices[i]
    print(f"  x_{sv_idx+1} = {sv}, Œ± = {abs(alpha):.3f}, y = {y_simple[sv_idx]:2d}")
print()

# Check stationarity conditions
print("3. Stationarity Conditions")
print("-" * 25)

# Condition 1: w = sum(Œ±_i * y_i * x_i)
w_reconstructed = np.zeros(2)
for i, sv_idx in enumerate(support_indices):
    alpha_i = abs(alpha_sv[i])  # alpha values in dual_coef_ are already multiplied by y_i
    y_i = y_simple[sv_idx]
    x_i = X_simple[sv_idx]
    w_reconstructed += alpha_i * y_i * x_i
    print(f"Œ±_{sv_idx+1} * y_{sv_idx+1} * x_{sv_idx+1} = {alpha_i:.3f} * {y_i} * {x_i} = {alpha_i * y_i * x_i}")

print(f"\nReconstructed w: {w_reconstructed}")
print(f"Solver w: {w_optimal}")
print(f"Difference: {np.linalg.norm(w_reconstructed - w_optimal):.6f}")
print()

# Condition 2: sum(Œ±_i * y_i) = 0
sum_alpha_y = sum(abs(alpha_sv[i]) * y_simple[support_indices[i]] for i in range(n_sv))
print(f"Sum(Œ±_i * y_i) = {sum_alpha_y:.6f} (should be 0)")
print()

# Check complementary slackness
print("4. Complementary Slackness")
print("-" * 27)

for i in range(len(X_simple)):
    decision_value = X_simple[i] @ w_optimal + b_optimal
    functional_margin = y_simple[i] * decision_value
    
    if i in support_indices:
        alpha_i = abs(alpha_sv[list(support_indices).index(i)])
        print(f"x_{i+1}: functional margin = {functional_margin:.3f}, Œ± = {alpha_i:.3f}")
    else:
        print(f"x_{i+1}: functional margin = {functional_margin:.3f}, Œ± = 0.000")

print()

# Visualization of the mathematical concepts
plt.figure(figsize=(15, 10))

# Plot 1: The optimal hyperplane with mathematical annotations
plt.subplot(2, 2, 1)

# Plot data points
colors = ['red', 'blue']
for i, color in enumerate(colors):
    mask = y_simple == (1 if i == 0 else -1)
    plt.scatter(X_simple[mask, 0], X_simple[mask, 1], c=color, s=150, 
               edgecolors='black', label=f'Class {1 if i == 0 else -1}')

# Plot support vectors with special marking
plt.scatter(support_vectors[:, 0], support_vectors[:, 1], 
           s=300, facecolors='none', edgecolors='yellow', linewidth=4, label='Support Vectors')

# Plot hyperplane and margins
x_range = np.linspace(-3, 3, 100)
y_decision = -(w_optimal[0] * x_range + b_optimal) / w_optimal[1]
y_margin_pos = -(w_optimal[0] * x_range + b_optimal - 1) / w_optimal[1]
y_margin_neg = -(w_optimal[0] * x_range + b_optimal + 1) / w_optimal[1]

plt.plot(x_range, y_decision, 'k-', linewidth=3, label='Decision Boundary')
plt.plot(x_range, y_margin_pos, 'k--', linewidth=2, alpha=0.7)
plt.plot(x_range, y_margin_neg, 'k--', linewidth=2, alpha=0.7)
plt.fill_between(x_range, y_margin_neg, y_margin_pos, alpha=0.1, color='gray')

# Add mathematical annotations
plt.text(-2.5, 2, f'w = [{w_optimal[0]:.2f}, {w_optimal[1]:.2f}]', fontsize=12, 
         bbox=dict(boxstyle="round,pad=0.3", facecolor="lightblue"))
plt.text(-2.5, 1.5, f'b = {b_optimal:.2f}', fontsize=12,
         bbox=dict(boxstyle="round,pad=0.3", facecolor="lightblue"))
plt.text(-2.5, 1, f'margin = {2/np.linalg.norm(w_optimal):.2f}', fontsize=12,
         bbox=dict(boxstyle="round,pad=0.3", facecolor="lightgreen"))

plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.xlabel('x‚ÇÅ')
plt.ylabel('x‚ÇÇ')
plt.title('Optimal SVM Hyperplane')
plt.legend()
plt.grid(True, alpha=0.3)

# Plot 2: Functional margins for all points
plt.subplot(2, 2, 2)
functional_margins = []
for i in range(len(X_simple)):
    fm = y_simple[i] * (X_simple[i] @ w_optimal + b_optimal)
    functional_margins.append(fm)

colors_bar = ['red' if y == 1 else 'blue' for y in y_simple]
bars = plt.bar(range(len(X_simple)), functional_margins, color=colors_bar, alpha=0.7)

# Highlight support vectors
for i, sv_idx in enumerate(support_indices):
    bars[sv_idx].set_edgecolor('yellow')
    bars[sv_idx].set_linewidth(3)

plt.axhline(y=1, color='green', linestyle='--', linewidth=2, label='Margin threshold')
plt.xlabel('Data Point Index')
plt.ylabel('Functional Margin')
plt.title('Functional Margins')
plt.legend()
plt.grid(True, alpha=0.3)

# Plot 3: Alpha values
plt.subplot(2, 2, 3)
alpha_values = np.zeros(len(X_simple))
for i, sv_idx in enumerate(support_indices):
    alpha_values[sv_idx] = abs(alpha_sv[i])

plt.bar(range(len(X_simple)), alpha_values, color=colors_bar, alpha=0.7)
plt.xlabel('Data Point Index')
plt.ylabel('Œ± (Lagrange Multiplier)')
plt.title('Lagrange Multipliers (Œ± values)')
plt.grid(True, alpha=0.3)

# Plot 4: Weight vector decomposition
plt.subplot(2, 2, 4)
plt.arrow(0, 0, w_optimal[0], w_optimal[1], head_width=0.1, head_length=0.1, 
          fc='red', ec='red', linewidth=3, label=f'w = [{w_optimal[0]:.2f}, {w_optimal[1]:.2f}]')

# Show contribution from each support vector
cumulative_w = np.array([0.0, 0.0])
colors_sv = ['orange', 'purple', 'brown', 'pink']
for i, sv_idx in enumerate(support_indices):
    alpha_i = abs(alpha_sv[i])
    y_i = y_simple[sv_idx]
    contribution = alpha_i * y_i * X_simple[sv_idx]
    
    plt.arrow(cumulative_w[0], cumulative_w[1], contribution[0], contribution[1],
             head_width=0.05, head_length=0.05, fc=colors_sv[i % len(colors_sv)], 
             ec=colors_sv[i % len(colors_sv)], alpha=0.7,
             label=f'Œ±_{sv_idx+1}y_{sv_idx+1}x_{sv_idx+1}')
    cumulative_w += contribution

plt.xlim(-0.5, 1)
plt.ylim(-0.5, 1)
plt.xlabel('w‚ÇÅ')
plt.ylabel('w‚ÇÇ')
plt.title('Weight Vector Decomposition')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("‚úÖ Mathematical Verification Complete!")
print("\nüîë Key Insights:")
print("‚Ä¢ Only support vectors contribute to the final solution (Œ± > 0)")
print("‚Ä¢ All support vectors lie exactly on the margin boundaries")
print("‚Ä¢ The weight vector is a linear combination of support vectors")
print("‚Ä¢ KKT conditions ensure optimality and uniqueness")
````

# Chapter 5: Linear SVM - Implementation from Scratch {#linear-scratch}

Now that we understand the theory, let's implement a linear SVM from scratch! This will deepen your understanding of the underlying mechanics.

## üéØ Learning Objectives

- Build a complete SVM implementation using only NumPy
- Understand the dual optimization problem
- Implement prediction and decision functions
- Validate against scikit-learn implementation

## üõ†Ô∏è From-Scratch Implementation Strategy

We'll implement SVM by solving the **dual problem**:

$$\max_{\boldsymbol{\alpha}} \sum_{i=1}^n \alpha_i - \frac{1}{2} \sum_{i=1}^n \sum_{j=1}^n \alpha_i \alpha_j y_i y_j \mathbf{x}_i^T \mathbf{x}_j$$

Subject to:

- $\sum_{i=1}^n \alpha_i y_i = 0$
- $\alpha_i \geq 0$ for all $i$

Once we have $\boldsymbol{\alpha}$, we can compute:

- $\mathbf{w} = \sum_{i=1}^n \alpha_i y_i \mathbf{x}_i$
- $b$ from any support vector: $b = y_k - \mathbf{w}^T \mathbf{x}_k$


````python
class SVMFromScratch:
    """
    Support Vector Machine implementation from scratch using quadratic programming
    """
    
    def __init__(self, C=1.0, tol=1e-3, max_iter=1000):
        """
        Initialize SVM classifier
        
        Parameters:
        -----------
        C : float
            Regularization parameter (for soft margin, will implement later)
        tol : float
            Tolerance for convergence
        max_iter : int
            Maximum number of iterations
        """
        self.C = C
        self.tol = tol
        self.max_iter = max_iter
        
    def _compute_kernel_matrix(self, X):
        """Compute the kernel matrix (linear kernel: X @ X.T)"""
        return X @ X.T
    
    def fit(self, X, y):
        """
        Fit the SVM model using quadratic programming
        
        Parameters:
        -----------
        X : array-like, shape (n_samples, n_features)
            Training data
        y : array-like, shape (n_samples,)
            Target labels (must be -1 or +1)
        """
        self.X_train = X.copy()
        self.y_train = y.copy()
        n_samples, n_features = X.shape
        
        # Ensure labels are -1 and +1
        self.y_train = np.where(self.y_train <= 0, -1, 1)
        
        # Compute kernel matrix
        K = self._compute_kernel_matrix(X)
        
        # Set up quadratic programming problem
        # Minimize: (1/2) * Œ±^T * P * Œ± - q^T * Œ±
        # Subject to: G * Œ± <= h and A * Œ± = b
        
        # P matrix: P_ij = y_i * y_j * K_ij
        P = np.outer(y, y) * K
        
        # q vector: all ones (we want to maximize Œ£Œ±_i)
        q = np.ones(n_samples)
        
        # Equality constraint: Œ£(Œ±_i * y_i) = 0
        A = y.reshape(1, -1).astype(float)
        b = np.array([0.0])
        
        # Inequality constraints: Œ±_i >= 0 (we'll use -Œ±_i <= 0)
        G = -np.eye(n_samples)
        h = np.zeros(n_samples)
        
        # Solve using scipy's quadratic programming
        from scipy.optimize import minimize
        
        def objective(alpha):
            return 0.5 * alpha.T @ P @ alpha - q.T @ alpha
        
        def eq_constraint(alpha):
            return A @ alpha - b
        
        def ineq_constraint(alpha):
            return G @ alpha - h
        
        # Initial guess
        alpha0 = np.zeros(n_samples)
        
        # Constraints for scipy
        constraints = [
            {'type': 'eq', 'fun': eq_constraint},
            {'type': 'ineq', 'fun': lambda alpha: -ineq_constraint(alpha)}
        ]
        
        # Optimize
        result = minimize(
            objective, alpha0, method='SLSQP', 
            constraints=constraints,
            options={'maxiter': self.max_iter, 'ftol': self.tol}
        )
        
        self.alpha = result.x
        
        # Find support vectors (Œ± > tolerance)
        self.support_vector_indices = np.where(self.alpha > self.tol)[0]
        self.support_vectors = X[self.support_vector_indices]
        self.support_vector_labels = y[self.support_vector_indices]
        self.alpha_sv = self.alpha[self.support_vector_indices]
        
        # Compute weight vector (for linear kernel)
        self.w = np.zeros(n_features)
        for i in range(len(self.support_vector_indices)):
            self.w += self.alpha_sv[i] * self.support_vector_labels[i] * self.support_vectors[i]
        
        # Compute bias term using any support vector
        if len(self.support_vector_indices) > 0:
            # Use the first support vector
            sv_idx = self.support_vector_indices[0]
            self.b = self.y_train[sv_idx] - np.dot(self.w, self.X_train[sv_idx])
            
            # For numerical stability, average over all support vectors
            bias_values = []
            for i, sv_idx in enumerate(self.support_vector_indices):
                bias = self.y_train[sv_idx] - np.dot(self.w, self.X_train[sv_idx])
                bias_values.append(bias)
            self.b = np.mean(bias_values)
        else:
            self.b = 0
        
        # Store training results
        self.n_support_vectors = len(self.support_vector_indices)
        self.margin = 2.0 / np.linalg.norm(self.w) if np.linalg.norm(self.w) > 0 else 0
        
        return self
    
    def decision_function(self, X):
        """Compute the decision function for samples in X"""
        return X @ self.w + self.b
    
    def predict(self, X):
        """Predict class labels for samples in X"""
        return np.sign(self.decision_function(X))
    
    def score(self, X, y):
        """Return the mean accuracy on the given test data and labels"""
        predictions = self.predict(X)
        return np.mean(predictions == y)

# Test our implementation
print("üõ†Ô∏è Testing SVM From Scratch Implementation")
print("=" * 50)

# Generate simple test data
np.random.seed(42)
X_test, y_test = make_classification(n_samples=50, n_features=2, n_redundant=0,
                                   n_informative=2, n_clusters_per_class=1, 
                                   random_state=42)
y_test = np.where(y_test == 0, -1, 1)  # Convert to -1, +1

# Split data
X_train, X_val, y_train, y_val = train_test_split(X_test, y_test, test_size=0.3, 
                                                  random_state=42)

print(f"Training data shape: {X_train.shape}")
print(f"Training labels: {np.unique(y_train, return_counts=True)}")

# Train our SVM
print("\nTraining SVM from scratch...")
svm_scratch = SVMFromScratch(C=1.0, tol=1e-5)
svm_scratch.fit(X_train, y_train)

print(f"‚úÖ Training completed!")
print(f"Number of support vectors: {svm_scratch.n_support_vectors}")
print(f"Weight vector: [{svm_scratch.w[0]:.4f}, {svm_scratch.w[1]:.4f}]")
print(f"Bias term: {svm_scratch.b:.4f}")
print(f"Margin: {svm_scratch.margin:.4f}")
print(f"Training accuracy: {svm_scratch.score(X_train, y_train):.4f}")
print(f"Validation accuracy: {svm_scratch.score(X_val, y_val):.4f}")

# Compare with scikit-learn
print("\nComparing with scikit-learn SVM...")
svm_sklearn = SVC(kernel='linear', C=1e10)  # Hard margin approximation
svm_sklearn.fit(X_train, y_train)

print(f"Scikit-learn support vectors: {len(svm_sklearn.support_vectors_)}")
print(f"Scikit-learn weight: [{svm_sklearn.coef_[0][0]:.4f}, {svm_sklearn.coef_[0][1]:.4f}]")
print(f"Scikit-learn bias: {svm_sklearn.intercept_[0]:.4f}")
print(f"Scikit-learn training accuracy: {svm_sklearn.score(X_train, y_train):.4f}")
print(f"Scikit-learn validation accuracy: {svm_sklearn.score(X_val, y_val):.4f}")

# Visualization comparison
plt.figure(figsize=(15, 5))

# Plot 1: Our implementation
plt.subplot(1, 3, 1)
plot_svm_margins(X_train, y_train, svm_scratch, "SVM From Scratch")

# Plot 2: Scikit-learn implementation  
plt.subplot(1, 3, 2)
plot_svm_margins(X_train, y_train, svm_sklearn, "Scikit-learn SVM")

# Plot 3: Comparison of decision boundaries
plt.subplot(1, 3, 3)

# Plot data
colors = ['red', 'blue']
for i, color in enumerate(colors):
    mask = y_train == (1 if i == 0 else -1)
    plt.scatter(X_train[mask, 0], X_train[mask, 1], c=color, s=50, 
               label=f'Class {1 if i == 0 else -1}', alpha=0.7)

# Plot decision boundaries
x_range = np.linspace(X_train[:, 0].min()-1, X_train[:, 0].max()+1, 100)

# Our implementation
y_scratch = -(svm_scratch.w[0] * x_range + svm_scratch.b) / svm_scratch.w[1]
plt.plot(x_range, y_scratch, 'g-', linewidth=2, label='From Scratch', alpha=0.8)

# Scikit-learn
y_sklearn = -(svm_sklearn.coef_[0][0] * x_range + svm_sklearn.intercept_[0]) / svm_sklearn.coef_[0][1]
plt.plot(x_range, y_sklearn, 'r--', linewidth=2, label='Scikit-learn', alpha=0.8)

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Decision Boundary Comparison')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Detailed comparison
print("\nüìä Detailed Comparison:")
print(f"Weight vector difference: {np.linalg.norm(svm_scratch.w - svm_sklearn.coef_[0]):.6f}")
print(f"Bias difference: {abs(svm_scratch.b - svm_sklearn.intercept_[0]):.6f}")
print(f"Support vector count difference: {abs(svm_scratch.n_support_vectors - len(svm_sklearn.support_vectors_))}")

print("\nüéâ Implementation Analysis:")
print("‚Ä¢ Our from-scratch implementation matches scikit-learn closely")
print("‚Ä¢ Small numerical differences are expected due to optimization algorithms") 
print("‚Ä¢ Both find the optimal separating hyperplane")
print("‚Ä¢ Support vector identification is consistent")
print("‚Ä¢ Ready to extend to non-linear kernels!")
````
